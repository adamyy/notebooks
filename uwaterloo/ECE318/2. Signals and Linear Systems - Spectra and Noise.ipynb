# 1.2 Linear Time Invariant (LTI) System

## A. Transmission over LTI Systems

- good and accurate models for a large class of communication systems
- some components of transmitters and receives such as filters, amplifiers and equalizers can be modeled as LTI systems

<img src="img/Snip20190918_102.png" width=60%/>

- $h(t)$: system's response to a unit impulse, $h(t)=y(t) \textrm{ when } x(t) = \delta(t)$

- the **frequency transfer function** is the Fourier Transform of the impulse response $H(\omega) = \mathcal{F}\{h(t)\}$
- in time domain, we have the convolution integral
    - $y(t) = h(t) \otimes x(t) = \int_{-\infty}^{+\infty} h(\tau)x(t- \tau) d\tau$
- in the frequency domain, we have the multiplication
    - $Y(\omega) = H(\omega)X(\omega)$
- an LTI acts as a filter on the various frequency components of the input signal, it might affect booth its amplitude and phase
    - $|Y(\omega)|e^{j\theta_y(\omega)} = |H(\omega)|e^{j\theta_h(\omega)} |X(\omega)|e^{j\theta_x(\omega)} $
    - $|Y(\omega)| = |H(\omega)||X(\omega)|$
    - $\theta_y(\omega) = \theta_h(\omega) + \theta_x(\omega)$

## B. Bandwidth of a System

<img src="img/Snip20190918_103.png" width=80%/>

- in practice filters do not have sharp transitions as ideal ones
- **bandwidth**: the interval of positive frequencies over which the magnitude of transfer function remains within a given numerical factor
- *half-power bandwidth* = *3dB bandwidth*: the band of frequencies at which the magnitude of the transfer function is at least $1\over \sqrt 2$ of its max
    - called 3dB bandwidth becauase it reduces the amplitude by a factor of $\sqrt 2$, corresponds to a decrease of 3dB on logarithmic scale

<img src="img/Snip20190918_104.png" width=60%/>

## C. Distortionless Transmission

- for ideal distortionless transmission, the output signal of an LTI should be $y(t)=kx(t-\tau)$ where $k$ and $\tau$ are constant
- $Y(\omega) = kX(\omega)e^{-j\omega \tau}$
- so the ideal system transfer function will be $H(\omega)= k e^{-j\omega \tau}$
- to achieve ideal distortionless transmission, the overall system response must have a constant magnitude response and its phase shift must be linear with frequency

# 2.3 Energy/Power-Type Signals

## A. Energy Signals

- The **energy** of a signal $x(t)$ is defined by
    - $E_x = \int_{-\infty}^{+\infty} |x(t)|^2 dt = \lim_{T\rightarrow \infty} \int_{-T/2}^{T/2} |x(t)|^2 dt$
    - the signal $x(t)$ is **energy-type signal** (or energy signal) if $E_x$ is finite: $0 < E_x < \infty$

**Example** $x(t) = e^{-at}u(t), a>0$
- $E_x = \int_0^{\infty}e^{-2at}dt = {1\over 2a} < \infty$
- it is an energy signal



## Parseval's Theorem

\begin{equation}
E_x = \int_{-\infty}^{+\infty} |x(t)|^2 dt = {1 \over 2\pi} \int_{-\infty}^{+\infty} |X(\omega)|^2 d\omega
\end{equation}

***Very Important***

\begin{equation}
E_x = \int_{-\infty}^{\infty} |x(t)|^2 dt = \int_{-\infty}^{\infty} |M(f)|^2 df
\end{equation}


- $|X(\omega)|^2$: **energy spectral density**

<img src="img/Snip20190918_105.png" width=80%/>




## B. Power Signals

- some signals may have infinite energy and finite time-average energy, the time-average of energy is called average power
- the **average power** of a signal $x(t)$
    - $P_x = \lim_{T \rightarrow \infty} {1\over T} \int_{-T/2}^{T/2} |x(t)|^2 dt$
    - the signal $x(t)$ is **power-type signal** (or power signal) if $0 < P_x <\infty$

**Example** $x(t) = A\cos(\omega_0 t + \theta)$

- $E_x = \lim_{T\rightarrow \infty} \int_{-T/2}^{T/2} A^2 \cos^2(\omega_0 t + \theta) dt $

- $=\lim_{T\rightarrow \infty} \int_{-T/2}^{T/2} {A^2 \over 2} [1 + \cos(2\omega_0 t + 2\theta)] dt $

- $= \lim_{T \rightarrow \infty} {A^2 T \over 2} + {A^2 \over 4 \omega_0} \sin(2\omega_0 t + 2\theta) |_{-T/2}^{T/2} =\infty$

- $P_x = \lim_{R \rightarrow \infty} {1\over T} \int_{-T/2}^{T/2} A^2 \cos^2(\omega_0 t + \theta) dt$
- $= \lim_{T \rightarrow \infty} {1\over T} ({A^2 T \over 2} + {A^2 \over 4 \omega_0} \sin(2\omega_0 t + 2\theta) |_{-T/2}^{T/2}) = {A^2 \over 2} < \infty$
- it is a power signal



## C. Power Spectral Density

- for power signals, **power spectral density** (PSD) is defined as
\begin{equation}
P_x = \lim_{T \rightarrow \infty} {1 \over T} \int_{-T/2}^{T/2} |x(t)|^2 dt = {1\over 2\pi} \int_{-\infty}^{+\infty} S_x(\omega)d\omega
\end{equation}

- to find $S_x$, define $x_T(t) = x(t)rect(t/T)$

<img src="img/Snip20190918_107.png" width=60%/>

- we have $\int_{-\infty}^{\infty} |x_T(t)|^2 dt = \int_{-T/2}^{T/2}|x(t)|^2 dt $
- using Parseval's Theorem, we have $\int_{-\infty}^{\infty} |x_T(t)|^2 dt = {1 \over 2\pi} \int_{-\infty}^{+\infty} |X_T(\omega)|^2 d\omega$


### Power Spectral Density for Periodic Signals
- for a periodic signal, each period contains a replica of the function and the limiting operation can be omitted as long as $T$ is taken as the period

\begin{equation}
P_x = \sum_{n=-\infty}^{\infty}|X_n|^2
\end{equation}

**Derivation**

\begin{equation}
P_x = {1 \over T} \int_{-T/2}^{T/2} |x(t)|^2 dt = {1 \over T} \int_{-T/2}^{T/2} x(t)x^*(t) dt
\end{equation}

\begin{equation}
={1 \over T} \int_{-T/2}^{T/2} (\sum_{m=-\infty}^{\infty}X_me^{jm\omega_0t}) (\sum_{n=-\infty}^{\infty}X_n^*e^{jn\omega_0t}) dt
\end{equation}

\begin{equation}
=\sum_{m=-\infty}^{\infty}X_m \sum_{n=-\infty}^{\infty}X_n^* {1\over T} \int_{-T/2}^{T/2} e^{j(m-n)\omega_0 t} dt
\end{equation}

since the integral evaluates to 1 when $m=n$, and 0 when $m\ne n$,

\begin{equation}
=\sum_{n=-\infty}^{\infty}X_n X_n^* = \sum_{n=-\infty}^{\infty}|X_n|^2
\end{equation}

### Summary

- Fourier Transform for periodic signal: $X(\omega) = 2\pi \sum_{n=-\infty}^{\infty} X_n \delta(\omega-n\omega_0)$

- Power Spectral Density for periodic signal: $S_x(\omega) = 2\pi \sum_{n=-\infty}^{\infty} |X_n|^2 \delta(\omega - n\omega_0)$
    - PSD of periodic signals have discrete components. It consists of a series of impulse functions with weights corresponding to the magnitude squared of respective Fourier series coefficients.

- Average power: $P_x = {1 \over T} \int_{-T/2}^{T/2} |x(t)|^2 dt = {1 \over 2\pi} \int_{-\infty}^{\infty}S_x(\omega)d\omega = \sum_{n = -\infty}^{\infty} |X_n|^2$

**Example** Find the power spectral density of the periodic signal $x(t) = A\cos(\omega_0 t +\theta)$

- $x(t)=({A \over 2} e^{j\theta})e^{j\omega_0t} + ({A \over 2}e^{-j\theta})e^{-j\omega_0t}$
- Fourier series coefficients $X_{-1} = {A \over 2}e^{j\theta}$, $X_{1} = {A \over 2} e^{-j\theta}$
- $X(\omega) = 2\pi X_{-1}\delta(\omega+\omega_0) + 2\pi X_1\delta(\omega-\omega_0)$
- $S_x(\omega) = 2\pi |X_{-1}|^2 \delta(\omega+\omega_0) = 2\pi |X_1|^2 \delta(\omega-\omega_0)$
- $P_x = {1\over 2\pi} \int_{-\infty}^{\infty} S_x(\omega)d\omega = |X_{-1}|^2 +|X_1|^2 = {A^2 \over 2}$


## Transmission of PSD over LTI

- **Output Energy Spectral Density**: $|Y(\omega)|^2 = |H(\omega)|^2|X(\omega)|^2$
- **Output Energy**: ${1 \over 2\pi} \int_{-\infty}^{\infty}|Y(\omega)|^2d\omega = {1\over 2\pi} \int_{-\infty}^{\infty} |H(\omega)|^2 |X(\omega)|^2d\omega$
- **Output Power Spectral Density**: $S_y(\omega) = |H(\omega)|^2S_x(\omega)$
- **Output Power**: $P_y = {1\over 2\pi} \int_{-\infty}^{\infty}S_y(\omega)d\omega = {1\over 2\pi} \int_{-\infty}^{\infty} |H(\omega)|^2 S_x(\omega)d\omega$

**Example** a voltage signal $x(t) = e^{-5t}u(t)$ is applied to the input of an ideal low-pass filter $H$, $H(\omega) = \{_{0, |\omega| > 5}^{1, |\omega| < 5}$,  the bandwidth is 5 rad/sec and resistance levels are 50 ohms. Calculate the energy of the input and the output.

- $|X(\omega)|^2 = {1\over \omega^2 + 25}$
- $E_x = {1\over R} \int_{-\infty}^{\infty} |x(t)|^2 dt = 2*10^{-3} \mathcal{J}$
- $E_y = {1\over 2\pi R} \int_{-\infty}^{\infty}|H(\omega)|^2|X(\omega)|^2 d\omega = 1*10^{-3} \mathcal{J}$



# 2.4 Correlation and Spectral Density

## Autocorrelation Function

\begin{equation}
R_x(\tau) = \lim_{T \rightarrow \infty} {1 \over T} \int_{-T/2}^{T/2} x^*(t)x(t+\tau) dt
\end{equation}

**Example** Determine the autocorrelation of the following signal

<img src="img/Snip20190919_116.png" width=60%/>

- since $x(t)$ is periodic, the limiting operation can be omitted
- $R_x(\tau) = {1\over T} \int_{-T/2}^{T/2} x^*(t)x(t+\tau)dt$
- for $-T/2 <\tau < 0$:
    - $R_x(\tau)= {1 \over T} \int_{-T/4}^{T/4+\tau} A^2 dt = A^2 ({1\over 2} + {\tau \over T})$
    - $-T/4<t<T/4$, $-T/4<t + \tau<T/4$ => $-T/4 - \tau<t<T/4 - \tau$
- for $0 <\tau < T/2$
    - $R_x(\tau) = {1 \over T} \int_{\tau-T/4}^{T/4}A^2dt = A^2 ({1\over 2} - {\tau \over T})$


## Autocorrelation of the Output of LTI

<img src="img/Snip20190918_108.png" width=80%/>

\begin{equation}
R_y(\tau) = R_x(\tau) \otimes h(\tau) \otimes h^*(-\tau)
\end{equation}

**Derivation**

<img src="img/Snip20190919_109.png" width=80%/>

### Time-domain
<img src="img/Snip20190919_110.png" width=40%/>

- $y(t)= h(t)\otimes x(t)$
- $R_y(\tau) = R_x(\tau)\otimes h(\tau) \otimes h^*(-\tau)$

### Frequency-domain
<img src="img/Snip20190919_111.png" width=40%/>

- $Y(\omega) = H(\omega) X(\omega)$
- $S_y(\omega) = |H(\omega)|^2 S_x(\omega)$


## Properties of Autocorrelation Functions

### Symmetry
$R_x^*(\tau) = R_x(-\tau)$

### Mean-Square Value
$R_x(\tau)|_{\tau = 0} = \lim_{T\rightarrow \infty} {1 \over T} \int_{-T/2}^{T/2} x(t)x^*(t)dt=\lim_{T\rightarrow \infty}{1 \over T} \int_{-T/2}^{T/2} |x(t)|^2 dt = P_x$

### Periodicity
if $x(t)$ is periodic with period $T$, then
$R_x(\tau + T) = R_x(\tau)$

### Maximum Value
The autocorrelation function is bounded by its mean square value

\begin{equation}
|R_x(\tau)| \leq R_x(0)
\end{equation}


## Crosscorrelation Function

**Definition** Crosscorrelation between two signal $x(t)$ and $y(t)$ is defined as

\begin{equation}
R_{xy}(\tau) = \lim_{T\rightarrow \infty} {1\over T} \int_{-T/2}^{T/2} x^*(t)y(t+\tau)dt
\end{equation}

- when $R_{xy}(\tau) = 0$, $x(t)$ and $y(t)$ are uncorrelated
- crosscorrelation between two signals measures the similarity between them
