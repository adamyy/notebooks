# 3.3 Angle Modulation

- amplitude-modulation methods are also called **linear-modulation methods**
    - although conventional AM is not linear

- **angle-modulation methods**: frequency modulation and phase modulation
- **frequency-modulation**: the frequency of the carrier $f_c$ is changed by the message signal
- **phase-modulation**: the phase of the carrier is changed according to the variations in the message signal

- angle-modulation methods are inherently non-linear, more complex to implement, and more difficult to analyze
- **bandwidth-expansion property**: frequency and phase modulation systems generally expand the bandwidth such that the effective bandwidth of the modulated signal is usually many times the bandwidth of the message signal
- **high degree of noise immunity**: FM systems are widely used in high-fidelity music broadcasting and point-to-point communication systems where the transmitter power is quite limited


## 3.3.1 Representation of FM and PM Signals

- an angle-modulated signal in general can be written as

\begin{equation}
u(t) = A_c \cos(\theta(t))
\end{equation}

- $\theta(t)$: the phase of the signal
- $f_i(t) = {1\over 2\pi} {d\over dt} \theta(t)$: instantaneous frequency 

- Since $u(t)$ is a bandpass signal, it can be represented as 
\begin{equation}
u(t) = A_c \cos(2\pi f_c t + \phi(t)) \\
f_i(t) = f_c + {1\over 2\pi} {d\over dt} \phi(t)
\end{equation}

### PM System

- **PM System**: $\phi(t) = k_p m(t)$
    - $k_p$: phase deviation constants
    - **maximum phase deviation in PM**: $\Delta \phi_{max} = k_p max(|m(t)|)$
    - **modulation index**: $\beta_p = k_p \max(|m(t)|)$
    - **instantaneous frequency**: $f_i(t) = f_c + k_p {1\over 2\pi} {d\over dt}m(t)$
    - *relationship between maximum phase deviation and modulation index*: $\beta_p = \Delta\phi_{max}$

### FM System

- **FM System**: $f_i(t) - f_c = k_f m(t) = {1\over 2\pi} {d \over dt} \phi(t)$
    - $\phi(t) = 2\pi k_f \int_{-\infty}^t m(\tau) d\tau$
    - $k_f$: frequency deviation constants
    - **maximum frequency deviation in FM**: $\Delta f_{max} = k_f max(|m(t)|)$
    - **modulation index**: $\beta_f = {k_f \max(|m(t)|) \over W}$, where $W$ denotes the **bandwidth of the message signal $m(t)$**
    - **instantaneous frequency**: $f_i(t) = f_c + k_f m(t)$
    - *relationship between maximum frequency deviation and modulation index*: $\beta_f = {\Delta f_{max} \over W}$

### For Message Signal $m(t) = a\cos(2\pi f_m t)$

- for a given message signal $m(t) = a\cos(2\pi f_m t)$
    - in PM we have $\phi(t) = k_p m(t) = k_p a \cos(2\pi f_m t)$
        - $u(t) = A_c \cos(2\pi f_c t + k_p a\cos(2\pi f_m t))$
        - modulation index $\beta_p = k_p a$

    - in FM we have $\phi(t) = 2\pi k_f \int_{n = -\infty}^{t} m(\tau)d\tau = {k_f a \over f_m} \sin(2\pi f_m t)$
        - $u(t) = A_c \cos(2\pi f_c t + {k_f a \over f_m}\sin(2\pi f_m t))$
        - modulation index $\beta_f = {k_f a \over f_m}$




### Relationship Between FM and PM

- if we phase modulate the carrier with the integral of a message, it is equivalent to frequency modulation of the carrier with the original message

\begin{equation}
{d\over dt} \phi(t) =
\begin{cases} 
      k_p {d\over dt} m(t), \textrm{PM} \\
      2\pi m(t), \textrm{FM}
\end{cases}
\end{equation}

- if we frequency modulate the carrier with the derivative of a message, the result is equivalent to phase modulation of the carrier with the message itself

### Demodulation of FM and PM 

- the demodulation of an FM signal involves finding the **instantaneous frequency** of the modulated signal and then subtracting the carrier frequency from it

- the demodulation of PM, the demodulation process is done by **finding the phase** of the signal and then recovering $m(t)$

# Narrowband Angle Modulation

- if in an angle-modulation system the deviation constants $k_p$ and $k_f$ and the message signal $m(t)$ are such that for all $t$ we have $\phi(t) << 1$, we have

\begin{equation}
u(t) = A_c \cos(2\pi f_c t) \cos(\phi(t)) - A_c \sin(2\pi f_c t) \sin(\phi(t)) \\
\approx A_c \cos(2\pi f_c t) - A_c \phi(t) \sin(2\pi f_c t)
\end{equation}

- modulated signal is *very similar to a conventional AM signal*
- the difference is that the message signal $m(t)$ is modulated ona sine carrier rather than a cosine carrier
- the bandwidth of this signal is similar to the bandwidth of a conventional AM signal, which is **approximately twice the bandwidth of the message**
- the narrowband angle-modulation method does not provide any better noise immunity compared to a conventional AM system
- thus seldom used in practice, but more in intermediate stage for generation of wideband angle-modulated signals



## 3.3.2 Spectral Characteristics of Angle-Modulated Signals

### Angle Modulation by a Sinusoidal Signal

\begin{equation}
u(t) = A_c \cos(2\pi f_c t + \beta \sin(2\pi f_m t))
\end{equation}

- $\beta$ is the modulation index that can be either $\beta_p$ or $\beta_f$, the modulated signal can also be written as 

\begin{equation}
u(t) = Re(A_c e^{j2\pi f_c t} e^{j\beta \sin(2\pi f_m t)})
\end{equation}

- $\sin(2\pi f_m t)$ is periodic with period $T_m = {1\over f_m}$, the same is true for the complex exponential signal $e^{j\beta \sin(2\pi f_m t)}$
- so it can be expanded in a Fourier series representation, the Fourier series coefficients are obtained from the integral

\begin{equation}
c_n = f_m \int_0^{1\over f_m} e^{j\beta \sin(2\pi f_m t)} e^{-jn2\pi f_m t} dt \\
= {1\over 2\pi} \int_0^{2\pi} e^{j(\beta\sin(u) - nu)} du
\end{equation}

- the latter integral is known as the **Bessel function of the first kind of order n** and is denoted by $J_n(\beta)$

- thus we have the Fourier series for the complex exponential as

\begin{equation}
e^{j\beta \sin(2\pi f_m t)} = \sum_{n = -\infty}^{\infty} J_n(\beta)e^{j2\pi nf_m t}
\end{equation}

\begin{equation}
u(t) = Re(A_c \sum_{n = -\infty}^{\infty} J_n(\beta)e^{j2\pi nf_m t} e^{j2\pi f_c t}) \\
=  \sum_{n = -\infty}^{\infty}  A_c J_n(\beta) \cos(2\pi (f_c + nf_m) t)
\end{equation}

- the angle-modulated signal contains all frequencies of the form $f_c + nf_m$ for $n = 0, \pm 1, \pm 2, ...$ therefore the actual bandwidth of the modulated signal is infinite; however, the amplitude of the sinusoidal components of frequencies $f_c \pm nf_m$ for large $n$ is very small, so we define a finite **effective bandwidth** for the modulated signal

#### Properties of Bessel function
\begin{equation}
J_n(\beta) = \sum_{k=0}^{\infty} {(-1)^k ({\beta \over 2})^{n+2k}\over k! (k+n)!}
\end{equation}

- for small $\beta$, $J_n(\beta) \approx {\beta^n \over 2^n n!}$
\begin{equation}
J_{-n}(\beta) = \begin{cases}
    J_n(\beta), \textrm{ for even n} \\
    -J_n(\beta), \textrm{ for odd n}
\end{cases}
\end{equation}


**Example 3.3.2** $c(t) = 10\cos(2\pi f_c t)$, $m(t) = \cos(20\pi t)$, assume FM with $k_f = 50$, find the expression for the modulated signal and determine how many harmonics should be selected to contain 99\% of the modulated signal power

- power content of the carrier signal $P_c = {A_c^2 \over 2} = 50$
- $u(t) = 10\cos(2\pi f_c t + 2\pi k_f \int_{-\infty}^t\cos(20\pi \tau) d\tau) = 10\cos(2\pi f_c t + 5 \sin(20\pi t))$
- modulation index $\beta = {k_f max(|m(t)|) \over W} = 50 \times 1 / 10 = 5 $
- thus the FM-modulated signal is $u(t) = \sum_{n = -\infty}^{\infty} 10 J_n(5) \cos(2\pi (f_c + 10n)t)$
- the frequency content of the modulated signal is concentrated at frequencies of the form $f_c + 10n$ for various $n$, to make sure that at least 99\% of the total power is within the effective bandwidth, we have to choose $k$ large enough such that
\begin{equation}
\sum_{n=-k}^{n = k} {1\over 2}A_c^2 J_n^2(5) \ge 0.99 \times 50
\end{equation}
- apply the symmetry properties of the Bessel function, we have
\begin{equation}
50(J_0^2(5) + 2\sum_{n=1}^k J_n^2(5)) \ge 49.5
\end{equation}




- in general the **effective bandwidth** of an angle-modulated signal, which contains at least 98\% of the signal power, is given by the relation $B_c = 2(\beta + 1)f_m$


### Angle Modulation by a Periodic Message Signal

- $m(t)$: arbitrary periodic message signal with period $T_m = {1\over f_m}$
- PM modulated signal: $u(t) = A_c \cos(2\pi f_c t + \beta m(t)) = A_c Re(e^{j2\pi f_c t} e^{j\beta m(t)})$
    - $e^{j\beta m(t)}$ will be periodic with the same period $T_m$
    - $e^{j\beta m(t)} = \sum_{n=-\infty}^{\infty} c_n e^{j2\pi n f_m t}$
    - $c_n = {1\over T_m} \int_0^{T_m} e^{j\beta m(t)} e^{-j2\pi n f_m t} dt = {1\over 2\pi} \int_0^{2\pi} e^{j(\beta m({u\over 2\pi f_m}) - nu)} du$
    
\begin{equation}
u(t) = A_c \sum_{n=-\infty}^{\infty} |c_n| \cos(2\pi (f_c + nf_m)t + \angle c_n)
\end{equation}

### Carson's Rule

- approximate the effective bandwidth of the modulated signal with

\begin{equation}
B_c = 2(\beta + 1)W
\end{equation}

## 3.3.3 Implementation of Angle Modulators

- any modulation and demodulation process involves the generation of new frequencies that were not present in the input signal -> for both amplitude and angle-modulation systems

- a modulator/demodulator cannot be modeled as a linear time-invariant system because a LTI system cannot produce any frequency components in the output that are not present in the input signal

- one method for generating an FM signal directly is to design an oscillator whose frequency changes with the input voltage
    - when the input voltage is zero, the oscillator generates a sinusoid with frequency $f_c$, and when the input voltage changes, this frequency changes accordingly 

- another approach for generating an angle-modulated signal is to first generate a NB signal, then change it to a WB signal (the *indirect method* for generation of FM/PM signals)

<img src="img/Snip20191111_7.png"/>

<img src="img/Snip20191111_8.png"/>

- to generate WB angle-modulated signals from NB angle-modulated signals:
    1. pass the NB signal to a frequency multiplier that multiplies the instantaneous frequency of the input by some constant $n$
        - usually done by applying the input signal to a nonlinear element then passing its output through a BPF tuned to the desired central frequency
        - input: $u_n(t) = A_c \cos(2\pi f_c t + \phi(t))$
        - output of the frequency multiplier: $y(t) = A_c \cos(2\pi n f_c t + n\phi(t))$
    2. since the carrier frequency $n_fc$ of the frequency multiplier output may not be desired carrier frequency, the last stage of the modulator performs an up/down conversion to shift the modulated signal to the desired center frequency
        - consists of a mixer and a BPF
        - if the frequency of the local oscillator of the mixer is $f_{L0}$ and we are using a down converter, the final wideband angle-modulated signal is given by 
        - $u(t) = A_c \cos(2\pi(nf_c - f_{L0})t + n\phi(t))$

## 3.3.3 Implementation of Angle Demodulators

<img src="img/Snip20191111_9.png"/>

- FM demodulators are implemented by generating an AM signal whose amplitude is proportional to the instantaneous frequency of the FM signal, then using an AM demodulator to recover the message signal

- to transform the FM signal into an AM signal, pass the FM signal through an LTI system whose frequency response is approximately a straight line in the frequency band of the FM signal
    - LTI system (example): $H(f) = V_0 + k(f-f_c)$ for $|f-f_c| < {B_c \over 2}$
    - input $u(t) = A_c \cos(2\pi f_c t + 2\pi k_f \int_{-\infty}^t m(\tau) d\tau)$
    - output: $v_o(t) = A_c (V_0 + k k_f m(t)) \cos(2\pi f_c t + 2\pi k_f \int_{-\infty}^t m(\tau) d\tau)$
    
### Feedback Demodulation

- another approach to FM demodulation is to use feedback in the demodulator to narrow the bandwidth of the FM detector and to reduce the noise power at the output of the demodulator


- **Phase-Locked Loop (PLL)**
<img src="img/Snip20191111_10.png"/>

- input: $u(t) = A_c\cos(2\pi f_c t + \phi(t))$
    - for FM: $\phi(t) = 2\pi k_f \int_{-\infty}^{t} m(\tau) d\tau$
- **VCO** generates a sinusoid of a fixed frequency, in this case the carrier frequency $f_c$, in the absence of an input control voltage
- let the output of the **loop filter** be $v(t)$
- then the instantaneous frequency of the VCO is
\begin{equation}
f_v(t) = f_c + k_v v(t)
\end{equation}

- $k_v$: deviation constant with units of Hz/volt, the VCO output may be expressed as
\begin{equation}
y_v(t) = A_v \sin(2\pi f_c t + \phi_v(t)) = A_v \sin(2\pi f_c t + 2\pi k_v \int_0^t v(\tau) d\tau)
\end{equation}

- the **phase comparator** is basically a **multiplier and filter** that rejects the signal component centered at $2f_c$
    - output: $e(t) = {1\over 2}A_v A_c \sin(\phi(t) - \phi_v(t))$
    - the difference $\phi(t) - \phi_v(t) = \phi_e(t)$ constitutes the phase error

- if we assume PLL is in lock, then phase error $\phi_e(t)$ is small, thus $\sin(\phi(t) - \phi_v(t)) \approx \phi(t) -\phi_v(t) = \phi_e(t)$

<img src="img/Snip20191111_11.png"/>

- under the above condition, we can use the linearized model of the PLL, in which $\phi_e(t) = \phi(t) - 2\pi k_v \int_0^t v(\tau) d\tau$
