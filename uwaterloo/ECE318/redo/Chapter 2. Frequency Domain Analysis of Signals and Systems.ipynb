# 2.1 Fourier Series

- Fourier series is applied to periodic signals whereas the Fourier transform can be applied to periodic and non-periodic signals


**Theorem 2.1.1.** (Fourier Series) Let the signal $x(t)$ be a periodic signal with period $T_0$, if the following conditions (the Dirichlet conditions) are satisfied
1. $x(t)$ is absolutely integrable over the period: $\int_0^{T_0} |x(t)|dt < \infty$
2. the number of maxima and minima of $x(t)$ in each period is finite
3. the number of discontinuities of $x(t)$ in each period is finite

then
\begin{equation}
x_{\pm}(t) = \sum_{n=-\infty}^{+\infty} x_n e^{j2\pi {n \over T_0} t} \\
x_n = {1\over T_0} \int_0^{T_0} x(t) e^{-j2\pi {n \over T_0} t} dt \\
x_{\pm}(t) = \begin{cases}
    x(t), \textrm{if continuous at t} \\
    {x(t^+) + x(t^-) \over 2}, \textrm{if discontinuous at t}
\end{cases}
\end{equation}

- $x_n$: Fourier series coefficients of $x(t)$
- $f_0 = {1\over T_0}$ is called the **fundamental frequency** of the signal $x(t)$
    - frequencies of the complex exponential signals are multiples of the fundamental frequency
    - the $n$th multiple of the fundamental frequency for positive $n$'s is called the **nth harmonic**


**Example 2.1.1** $x(t) = \sum_{n=-\infty}^{\infty} \Pi({t - nT_0 \over \tau})$ periodic rectangular pulse, find Fourier series

\begin{equation}
x_n = {1\over T_0} \int_{-T_0 / 2}^{T_0 / 2} x(t) e^{-jn {2\pi \over T_0} t} dt \\
= {1\over T_0} {T_0 \over -jn2\pi} (e^{-jn {\pi \tau \over T_0}} - e^{+jn {\pi \tau \over T_0}}) = {1\over \pi n} \sin({n\pi \tau \over T_0}) = {\tau \over T_0} sinc({n\tau \over T_0}) \\
x(t) = \sum_{n=-\infty}^{+\infty} {\tau \over T_0} sin({n\tau \over T_0}) e^{jn {2\pi t \over T_0}}
\end{equation}




## 2.1.1 Trigonometric Fourier Series

\begin{equation}
x_n = {1\over T_0} \int_{T_0} x(t) e^{-j2\pi {n \over T_0} t} dt = {a_n \over 2} - j{b_n \over 2} \\
a_n = {2\over T_0} \int_{T_0} x(t) \cos(2\pi {n \over T_0} t) dt \\
b_n = {2\over T_0} \int_{T_0} x(t) \sin(2\pi {n \over T_0} t) dt
\end{equation}


\begin{equation}
x(t) = \sum_{n=-\infty}^{+\infty} x_n e^{j2\pi {n \over T_0} t}
= x_0 + 2 \sum_{n=1}^{\infty} |x_n| \cos(2\pi {n \over T_0} t + \angle{x_n}) \\
= {a_0 \over 2} + \sum_{n=1}^{\infty} (a_n \cos(2\pi {n \over T_0} t) + b_n \sin(2\pi {n \over T_0}t))
\end{equation}


\begin{equation}
|x_n| = {1\over 2} \sqrt{a_n^2 + b_n^2} \\
\angle{x_n} = -arctan({b_n \over a_n})
\end{equation}


# 2.2 Fourier Transforms

- extension of the Fourier series to the general class of signals (periodic and nonperiodic)

**Theorem 2.2.1** (Fourier Transform) If the signal $x(t)$ satisfies certain conditions known as the Dirchlet conditions, the Fourier transform of $x(t)$ is defined as

\begin{equation}
X(f) = \mathcal{F}\{x(t)\} = \int_{-\infty}^{\infty} x(t) e^{-j2\pi f t} dt \\
x(t) = \mathcal{F}^{-1}\{X(f)\}  = \int_{-\infty}^{\infty} X(f) e^{j2\pi f t} df
\end{equation}

- the function $X(f)$ is sometimes referred to as the spectrum of the signal $x(t)$

**Example 2.2.1** find Fourier transform of $\Pi(t)$

\begin{equation}
\mathcal{F}(\Pi(t)) = \int_{-\infty}^{\infty} \Pi(t) e^{-j2\pi f t} dt\\
= \int_{-1/2}^{+1/2} e^{-j2\pi f t} dt = {1\over -j2\pi f} (e^{-j\pi f} - e^{j\pi f}) = {\sin(\pi f) \over \pi f} = sinc(f)
\end{equation}


**Example 2.2.2** find Fourier transform of $x(t) = \delta(t)$

\begin{equation}
\mathcal{F}(\delta(t)) = \int_{-\infty}^{\infty} \delta(t) e^{-j2\pi f t} dt = 1
\end{equation}

- the **sifting property** of $\delta(t)$: $\int_{-\infty}^{\infty} x(t) \delta(t - t_0) dt = x(t_0)$

## 2.2.1 Fourier Transform of Real, Even and Odd Signals

\begin{equation}
\mathcal{F}(x(t)) = \int_{-\infty}^{+\infty} x(t) e^{-j2\pi f t} dt \\
= \int_{-\infty}^{+\infty} x(t) \cos(2\pi f t) dt - j \int_{-\infty}^{+\infty} x(t) \sin(2\pi f t) dt
\end{equation}

- Hermitian function: $X(-f) = X^*(f)$

## 2.2.2 Basic Properties of the Fourier Transform

- **linearity**
    - if $x(t) => X(f), y(t) => Y(f)$
    - then $ax(t) + by(t) => aX(f)+bY(f)$
- **duality**
    - if $X(f) = \mathcal{F}(x(t))$, then $x(f) = \mathcal{F}(X(-t))$ and $x(-f) = \mathcal{F}(X(t))$
- **time-shift**
    - a shift of $t_0$ in the time origin causes a phase shift of $-2\pi f t_0$ in the frequency domain
    - $\mathcal{F}(x(t-t_0)) = e^{-j2\pi f t_0} \mathcal{F}(x(t))$
- **scaling**
    - $\mathcal{F}(x(at)) = {1\over |a|} X({f \over a})$
- **convolution**
    - $\mathcal{F}(x(t) \otimes y(t)) = \mathcal{F}(x(t)) * \mathcal{F}(y(t)) = X(f) Y(f)$
- **modulation**
    - $\mathcal{F}(x(t)e^{j2\pi f_0 t}) = X(f - f_0)$
    - $\mathcal{F}(x(t)\cos(2\pi f_0 t)) = {1\over 2} X(f - f_0) + {1\over 2} X(f + f_0)$
- **Parseval**
    - $\int_{-\infty}^{\infty} x(t)y^*(t) dt = \int_{-\infty}^{\infty}X(f)Y^*(f) df$
- **Rayleigh**
    - $\int_{-\infty}^{\infty} |x(t)|^2 dt = \int_{-\infty}^{\infty} |X(f)|^2 df$
- **Autocorrelation**
    - time autocorrelation function $R_x(\tau) = \int_{-\infty}^{\infty} x(t)x^*(t-\tau)dt$
    - $\mathcal{F}(R_x(\tau)) = |X(f)|^2$
- **Differentiation**
    - $\mathcal{F}({d \over dt}x(t)) = j2\pi f X(f)$
- **Integration**
    - $\mathcal{F}(\int_{-\infty}^t x(\tau) d\tau) = {X(f) \over j2\pi f} + {1\over 2} X(0)\delta(f)$
- **Moments**
    - the nth moment of $x(t)$: $\int_{-\infty}^{\infty} t^n x(t) dt = ({j \over 2\pi}) ^n {d^n \over df^n} X(f) |_{f = 0}$

## 2.2.3 Fourier Transform for Periodic Signals

- let $x(t)$ be a periodic signal with period $T_0$, satisfying the Dirichlet conditions, let $x_n$ denote the Fourier series coefficients corresponding to this signal, then
\begin{equation}
x(t) = \sum_{n = -\infty}^{\infty} x_n e^{j2\pi {n \over T_0}t}
\end{equation}

- taking the Fourier transform of both sides and using the fact that

\begin{equation}
\mathcal{F}(e^{j2\pi {n \over T_0}t}) = \delta(f - {n \over T_0})
\end{equation}

- we have 

\begin{equation}
X(f) = \sum_{n=-\infty}^{\infty} x_n \delta(f - {n \over T_0})
\end{equation}

- the Fourier transform of a periodic signal $x(t)$ consists of a sequence of impulses in frequency at multiples of the fundamental frequency of the periodic signal

- Given the periodic signal $x(t)$, we can find $x_n$ 
    1. first find the truncated signal $x_{T_0}(t)$
        - $x_{T_0}(t) = \begin{cases}
            x(t), {-T_0 \over 2} < t \le {T_0 \over 2} \\
            0, \textrm{otherwise}
        \end{cases}$
    2. then determine the Fourier transform of the truncated signal
    3. evaluate the Fourier transform of the truncated signal at $f = {n \over T_0}$, to obtain the nth harmonic and multiply by ${1\over T_0}$
    
**Example 2.2.3** find the Fourier series coefficients of the periodic gate function $x(t) = \sum_{n=-\infty}^{\infty} \Pi({t - nT_0 \over \tau})$, where $\tau$ is a constant

- truncated signal $x_{T_0}(t) = \Pi({t \over \tau})$
- fourier transform $X_{T_0}(f) = \tau sinc(\tau f)$
- therefore, $x_n = {\tau \over T_0} sinc({n\tau \over T_0})$





# 2.3 Power and Energy

**energy**: $E_x = \sum_{-\infty}^{\infty} |x(t)|^2 dt$
- a signal is energy-type if $E_x < \infty$


**power content**: $P_x = \lim_{T \rightarrow \infty} {1\over T} \int_{-T \over 2}^{+T \over 2} |x(t)|^2 dt$
- a signal is power-type if $0 < P_x < \infty$
- periodic signals (except for $x(t)=0$) are power type and have power $P_x = {1\over T_0} \int_{T_0} |x(t)|^2 dt$


- a signal cannot be both power and energy type because energy type signals have power content $P_x = 0$ and power type signals have energy of $E_x = \infty$

- most of the signals are either energy-type or power type




## 2.3.1 Energy-Type Signals

- for an energy-type signal $x(t)$, the autocorrelation function $R_x(\tau)$ is defined as
\begin{equation}
R_x(\tau) = x(\tau) \otimes x^*(-\tau) = \int_{-\infty}^{\infty} x(t) x^*(t - \tau) dt
\end{equation}

- by setting $\tau = 0$, we get the energy content of $x(t)$
    - $\mathcal{E}_x = R_x(0) = \int_{-\infty}^{\infty} |x(t)|^2 dt$
    - by Rayleigh's theorem, $\mathcal{E}_x = \int_{-\infty}^{\infty} |X(f)|^2 df$

- there are two methods for finding the energy in a signal, one uses $x(t)$, another uses $X(f)$


**energy spectral density** $\mathcal{G}_x(f) = \mathcal{F}(R_x(\tau)) = |X(f)|^2$
- represents the amount of energy per hertz of bandwidth present in the signal at various frequencies


**energy content** $\mathcal{E}_x = R_x(0) = \int_{-\infty}^{\infty} \mathcal{G}_x(f) df = \int_{-\infty}^{\infty} |X(f)|^2 df$


**filtered energy**: $y(t) = x(t) \otimes h(t)$, then $R_y(\tau) = R_x(\tau) \otimes R_h(\tau)$, then $\mathcal{G}_y(f) = \mathcal{G}_x(f) \mathcal{G}_h(f) = |H(f)|^2 |X(f)|^2$



## 2.3.2 Power-Type Signals

**time-average autocorrelation function** of power-type signal $x(t)$
\begin{equation}
R_x(\tau) = \lim_{T->\infty} {1\over T} \int_T x(t) x^*(t - \tau) dt
\end{equation}
    

**power content**

\begin{equation}
P_x = lim_{T->\infty} {1\over T} \int_T |x(t)|^2 dt = R_x (0)
\end{equation}


**power-spectral density / power spectrum** is defined as the Fourier transform of the time-average autocorrelation function
\begin{equation}
S_x(f) = \mathcal{F}(R_x(\tau))
\end{equation}


**power content in terms of power spectral density**
\begin{equation}
P_x = R_x(0) = \int_{-\infty}^{\infty} S_x(f) df
\end{equation}


**filtered power**
\begin{equation}
S_y(f) = S_x(f) |H(f)|^2
\end{equation}

# 2.6 Hilbert Transform

\begin{equation}
\hat{x}(t) = {1\over \pi t} \otimes x(t) = {1\over \pi} \int_{-\infty}^{\infty} {x(t) \over t - \tau} d\tau \\
\mathcal{F}(\hat{x}(t)) = -j sgn(f) X(f)
\end{equation}

- in positive frequencies, the spectrum of the signal is multiplied by $-j$, in negative frequencies, it is multiplied by $+j$


**Example** x(t) = 2sinc(2t)

- $\mathcal{F}(x(t)) = 2 \times {1\over 2} \Pi({f\over 2}) = \Pi({f\over 2}) = \Pi(f +{1\over 2}) + \Pi(f - {1\over 2})$
- $\mathcal{F}(\hat{x}(t)) = j\Pi(f+{1\over 2}) - j\Pi(f - {1\over 2})$
- $\hat{x}(t) = je^{-j\pi t} sinc(t) - je^{j\pi t} sinc(t) = -j(e^{j\pi t} - e^{-j\pi t}) sinc(t) = 2\sin(\pi t) sinc(t)$

- **evenness and oddness**: the Hilbert transform of an even signal is odd, that of an odd signal is even

- **sign reversal**: apply the Hilbert transform twice causes a sign reversal of the signal
\begin{equation}
\hat{\hat{x}}(t) = -x(t)
\end{equation}

- **energy**: the energy content of a signal is equal to the energy content of its Hilbert transform

- **orthogonality**: the signal $x(t)$ and its Hilbert transform are orthogonal: $\int_{-\infty}^{\infty} x(t)\hat{x}(t) dt = 0$


# 2.7 Lowpass and Bandpass Signals

- a **lowpass** signal is a signal in which the spectrum (frequency content) of the signal is located around the zero frequency
- a **bandpass** signal is a signal with a spectrum far from the zero frequency
    - the frequency spectrum of a bandpass signal is usually located around a frequency $f_c$ which is much higher than the bandwidth of the signal 

- given $x(t) = x_c \cos(2\pi f_c t) - x_s \sin(2\pi f_c t)$
    - the **in-phase component** is $x_c$, which is in the direction of $\cos(2\pi f_c t)$
    - the **quadrature component** is $x_s$, which is in the direction of $-\sin(2\pi f_c t)$
    - both the in-phase and quadrature components of a bandpass signal are lowpass signals
    - a bandpass signal can be represented in terms of two lowpass signals, namely its in-phase and quadrature components
    - the complex lowpass signal $x_l(t) = x_c(t) +j x_s(t)$ is called the **lowpass equivalent** of the bandpass signal $x(t)$
    - $x(t) = Re(x_l(t) e^{j2\pi f_c t})$
