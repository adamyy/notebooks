# Analog Communication System

<img src="img/Snip20190925_8.png" width=80%/>

- analog may be transmitted directly via carrier modulation over the propagation channel and to be carrier-demodulated at the receiver
    - transmitter -> modulator
    - receiver -> demodulator

- **modulation**: the process by which some characteristics of a carrier signal (i.e., modulated signal) is varied in accordance with message signal (i.e., modulating signal)

- $f(t)$: message signal (*baseband* signal)
    - a bandlimited signal whose frequency content is in the neighborhood of $f=0$ (DC) is reffered to as a baseband signal

- $c(t)$: the carrier signal, independent of $f(t)$
    - $c(t) = A_c \cos(2\pi f_c t + \theta_c)$
    - $A_c$: carrier amplitude
    - $f_c$: carrier frequency, $\omega_c = 2\pi f_c$
    - $\theta_c$: carrier phase
- $f(t)$ modulates $c(t)$ in either amplitude, frequency or phase. In effect, modulation converts $f(t)$ to a *passband* (or bandpass) form, in the neighborhood of the center frequency $f_c$

- why is modulation required
    - to accommodate for simultaneous transmission of several signals (e.g., radio/TV)
    - to expand the bandwidth of the transmitted signal for better transmission quality (to reduce interference)
        - Shannon Theorem: $C = B log_2(1+SNR)$
            - $C$: channel capacity, maximum achievable information rate that can be transmitted over the channel
            - $B$: bandwidth
            - $SNR$: Signal-to-noise ratio


# Double Side Band Large Carrier (DSB-LC)

<img src="img/Snip20190925_9.png" width=80%/>

\begin{equation}
F(\omega) = \mathcal{F}\{f(t)\}
\end{equation}

\begin{equation}
\Phi(\omega) = \mathcal{F}\{A_c\cos(\omega_c t) + f(t)\cos(\omega_c t)\}
\end{equation}
\begin{equation}
= \mathcal{F}\{ {A_c \over 2} e^{j\omega_c t} + {A_c \over 2} e^{-j\omega_c t} + {f(t) \over 2} e^{j\omega_c t} + {f(t) \over 2} e^{-j\omega_c t} \}
\end{equation}
\begin{equation}
= \pi A_c \delta(\omega-\omega_c) + \pi A_c \delta(\omega+\omega_c) + {1\over 2}F(\omega-\omega_c) + {1\over 2} F(\omega+\omega_c)
\end{equation}

<img src="img/Snip20190925_10.png" width=80%/>

<img src="img/Snip20190925_11.png" width=80%/>

- modulation shifts the content of $F(\omega)$ to the neighborhood of $\omega_c = 2\pi f_c$
- **lower sideband**: $F(\omega) \textrm{ for } \omega\in [-2\pi B, 0]$ is shifted to $\Phi(\omega)$ for $\omega \in [\omega_c - 2\pi B, \omega_c]$
- **upper sideband**: $F(\omega) \textrm{ for } \omega\in [0, 2\pi B]$ is shifted to $\Phi(\omega)$ for $\omega \in [\omega_c, \omega_c + 2\pi B]$
- Let $B$ denote the highest frequency component of $f(t)$, assume $f_c >> B$, then $\phi(t)$ is defined as a *narrowband* signal (i.e., its spectral content is located in the immediate vicinity of some high center frequency)
- The **bandwidth** of the message signal is $B$, the transmission bandwidth $\beta_T = 2B$, thus DSB-LC is wasteful of bandwidth

### Envelope Distortion and Over Modulation

<img src="img/Snip20190925_12.png" width=80%/>

- if $A_c + f(t) > 0$ for all $t$, the envelope of $\phi(t)$ has essentially the same shape as $f(t)$
- if $A_c + f(t) < 0$ for any $t$, the carrier wave becomes **over-modulated**, resulting in carrier phase reversal whenever $A_c + f(t)$ crosses zero, resulting envelope distortion in $\phi(t)$

### DSB-LC Modulation Index

- to evaluate the performance of DSB-LC modulation, the modulation index is introduced

\begin{equation}
f(t) = A_m \cos(\omega_m t), c(t) = A_c \cos(\omega_c t), \phi(t) = (A_c + A_m \cos(\omega_m t)) \cos(\omega_c t)
\end{equation}

- modulation index $\alpha$: $\alpha = A_m / A_c$
- $\phi(t) = A_c(1+\alpha \cos(\omega_m t)) \cos(\omega_c t)$

- Let $A_{max}$ and $A_{min}$ denote the maximum and minimum values of the envelope of $\phi(t)$, then
    - $A_{max} = A_c(1+\alpha), A_{min} = A_c(1-\alpha)$
    - ${A_{max} \over A_{min}} = {1 + \alpha \over 1 - \alpha}$
    - $\alpha = {A_{max} - A_{min} \over A_{max} + A_{min}}$

**Remark** Note that $\cos(\omega_m t)\cos(\omega_c t) = {1\over 2} \cos(\omega_c + \omega_m)t + {1\over 2} \cos(\omega_c - \omega_m)t$

- the modulated signal can be written as
\begin{equation}
\phi(t) = A_c \cos(\omega_c t) + {\alpha \over 2} A_c \cos(\omega_c + \omega_m)t + {\alpha \over 2} A_c \cos(\omega_c - \omega_m)t 
\end{equation}
- which has three different frequencies


- for a general message, to prevent overmodulation, we put
\begin{equation}
\phi(t) = A_c (1+ \alpha f_n(t))\cos(\omega_c t) \textrm{, where } f_n(t) = {f(t) \over \max|f(t)|}
\end{equation}
    - $0 < \alpha < 1$ is the modulation index so chosen to prevent envelope zero-crossing

### Carrier and Sideband Power in DSB-LC

$\phi(t) = A_c(1+\alpha \cos(\omega_m t)) \cos(\omega_c t)$

$\overline{\phi^2(t)} = A_c^2 \overline{\cos^2(\omega_ct)} + \overline{f^2(t)\cos^2(\omega_ct)} + 2 A_c \overline{f(t)cos^2(\omega_c t)}$

Assume $\overline{f(t)} = 0$ and $f(t)$ varies slowly with respect to $\cos(\omega_c t)$ which implies the last term of the above identity is zero.

$\overline{\phi^2(t)} = A_c^2 \overline{\cos^2(\omega_ct)} + \overline{f^2(t)}\textrm{ }\overline{\cos^2(\omega_ct)}$

$\overline{\cos^2(\omega_ct)} = \lim_{T\rightarrow \infty} {1 \over T} \int_{-T/2}^{T/2} \cos^2(\omega_c t) dt = \lim_{T\rightarrow \infty} {1 \over T} \int_{-T/2}^{T/2} {1\over 2}(1+\cos(2\omega_c t)) dt = 1/2$

Thus
\begin{equation}
\overline{\phi^2(t)} = A_c^2 / 2 + \overline{f^2(t)} / 2
\end{equation}

- carrier power: $A_c^2 / 2$
- sideband power (carries information): $\overline{f^2(t)} / 2$
- modulation efficiency: $\mu = {\textrm{useful power} \over \textrm{total power}} = {\overline{f^2(t)} / 2 \over A_c^2 / 2 + \overline{f^2(t)} / 2}$




**Example** Given the example of single-frequency sinusoidal signal ($A_m \cos(\omega_m t)$), find the DSB-LC signal, the power in its upper and lower sidebands, and modulation efficiency, assuming the modulation index is $\alpha$.

\begin{equation}
\phi(t) = A_c(1+\alpha \cos(\omega_m t)) \cos(\omega_c t) \\
= A_c \cos(\omega_c t) + {\alpha A_c \over 2} \cos(\omega_c + \omega_m)t + {\alpha A_c \over 2} \cos(\omega_c - \omega_m)t
\end{equation}

- upper sideband power: $({\alpha A_c \over 2})^2 \overline{\cos^2(\omega_c + \omega_m)t} = {\alpha^2 A_c^2 \over 8}$
- lower sideband power: $({\alpha A_c \over 2})^2 \overline{\cos^2(\omega_c - \omega_m)t} = {\alpha^2 A_c^2 \over 8}$
- modulation efficiency $\mu = {useful \over total} = {{\alpha^2 A_c^2 \over 8} + {\alpha^2 A_c^2 \over 8} \over {\alpha^2 A_c^2 \over 8} + {\alpha^2 A_c^2 \over 8} + A_c^2 / 2} = {\alpha^2 \over 2 + \alpha^2}$

# Double Side Band Suppressed Carrier (DSB-SC)

### Modulation of DSB-SC

<img src="img/Snip20191008_136.png" width=60%/>

\begin{equation}
\Phi(\omega) = \mathcal{F}\{A_c f(t)\cos(\omega_c t)\} \\
= \mathcal{F}\{{A_c \over 2} f(t)e^{j\omega_c t} + {A_c \over 2} f(t) e^{-j\omega_c t} \} \\
= {A_c \over 2}F(\omega-\omega_c) + {A_c \over 2}F(\omega + \omega_c)
\end{equation}

<img src="img/Snip20191008_137.png" width=80%/>

- $\phi(t)$ undergoes a phase reversal whenever $f(t)$ crosses zero, the envelope of $\phi(t)$ is different from $f(t)$, both amplitude and phase of $\phi(t)$ carry information of $f(t)$
- the transmission bandwidth required by DSB-SC is the same as that for DSB-LC, $\beta_T = 2B$

### Demodulation of DSB-SC

<img src="img/Snip20191008_138.png" width=80%/>

\begin{equation}
x(t) = \phi(t)A_c'\cos(\omega_c t) = A_c'A_c f(t)\cos^2(\omega_c t) = {1\over 2} A_c'A_c f(t) + {1\over 2}A_c'A_c f(t)\cos(2\omega_c t)
\end{equation}

\begin{equation}
X(\omega) = {1\over 2} A_c A_c'F(\omega)  + {1\over 4}A_c A_c'[F(\omega-2\omega_c) + F(\omega+2\omega_c)]
\end{equation}

#### Error at the Receiver

<img src="img/Snip20191008_140.png" width=70%/>

<img src="img/Snip20191008_141.png" width=70%/>

\begin{equation}
z(t) = {1\over 2}A_c A_c'f(t)\cos(\Delta\omega t + \theta_0)
\end{equation}

- if $\Delta\omega \ne 0$, $z(t) = {1\over 2}A_c A_c'f(t)\cos(\Delta\omega t)$
- if $\theta_0 \ne 0$, $z(t) = {1\over 2}A_c A_c'f(t)\cos(\theta_0)$

- to recover $f(t)$ accurately from $\phi(t)$, we need to use a synchronized oscillator (coherent detection, synchronous detection required)
