- an angle-modulated signal in general can be written as

\begin{equation}
u(t) = A_c \cos(\theta(t))
\end{equation}

- $\theta(t)$: the phase of the signal
- $f_i(t) = {1\over 2\pi} {d\over dt} \theta(t)$: instantaneous frequency 

- $u(t)$ is a bandpass signal, it can be represented as 
\begin{equation}
u(t) = A_c \cos(2\pi f_c t + \phi(t)) \\
f_i(t) = f_c + {1\over 2\pi} {d\over dt} \phi(t)
\end{equation}

- **PM System**: $\phi(t) = k_p m(t)$
    - $k_p$: phase deviation constants
- **FM System**: $f_i(t) - f_c = k_f m(t) = {1\over 2\pi} {d \over dt} \phi(t)$
    - $k_f$: frequency deviation constants

<img src="img/Snip20191110_218.png" width=60%/>

- **maximum phase deviation in PM**: $\Delta \phi_{max} = k_p max(|m(t)|)$
- **maximum frequency deviation in FM**: $\Delta f_{max} = k_f max(|m(t)|)$


- for a given message signal $m(t) = a\cos(2\pi f_m t)$
    - in PM we have $\phi(t) = k_p m(t) = k_p a \cos(2\pi f_m t)$
        - $u(t) = A_c \cos(2\pi f_c t + k_p a\cos(2\pi f_m t))$
        - modulation index $\beta_p = k_p a$

    - in FM we have $\phi(t) = 2\pi k_f \int_{n = -\infty}^{t} m(\tau)d\tau = {k_f a \over f_m} \sin(2\pi f_m t)$
        - $u(t) = A_c \cos(2\pi f_c t + {k_f a \over f_m}\sin(2\pi f_m t))$
        - modulation index $\beta_f = {k_f a \over f_m}$


- **for a general signal $m(t)$**
    - $\beta_p = k_p \max(|m(t)|)$
    - $\beta_f = {k_f \max(|m(t)|) \over W}$, where $W$ denotes the **bandwidth of the message signal $m(t)$**

- **in terms of the maximum phase and frequency deviation** $\Delta\phi_{max}$ and $\Delta f_{max}$
    - $\beta_p = \Delta\phi_{max}$
    - $\beta_f = {\Delta f_{max} \over W}$



# Angle Modulation

- **Amplitude modulation (AM)**: the amplitude of the carrier varies in accordance with the message signal $f(t)$
    - the spectrum of the modulated signal is shifted/scaled version of the message signal spectrum
    - transmission bandwidth vs message bandwidth: $B \le B_T \le 2B$
    - the signal to noise ratio (SNR) can be increased only by increasing the transmitted signal power
- **Angle modulation (FM, PM)**: the angle of the carrier is varied in accordance with the message signal while the amplitude of the carrier is constant
    - the spectrum of the modulated signal is NOT simply related to message signal spectrum
    - transmission bandwidth >> message bandwidth: the **bandwidth-expansion property**
        - both FM and PM systems generally expand the bandwidth such that the effective bandwidth of the modulated signal is usually many times the bandwidth of the message signal
    - improvement in SNR without increasing the transmitted signal power
        - the major benefit of these systems is their high degree of noise immunity
        - trade-off bandwidth for high noise immunity
    - angle-modulation methods are nonlinear
    - FM systems are widely used in high-fidelity music broadcasting and point-to-point communication systems where transmitter power is quite limited

- An angle-modulated signal in general form

\begin{equation}

\end{equation}

- $\theta(t)$ is the phase of the signal, and its instantaneous frequency $f_i(t)$ is given by 

\begin{equation}
f_i(t) = {1\over 2\pi} {d\over dt} \theta(t)
\end{equation}

- since $u(t)$ is a bandpass signal, it can be represented as 

\begin{equation}

\end{equation}

## Frequency Modulation (FM)
- the frequency of the carrier $f_c$ is changed by the message signal

\begin{equation}
u(t) = A_c\cos(2\pi f_c t + \phi(t))
\end{equation}

- for FM system,

\begin{equation}
f_i(t) - f_c = k_f m(t) = {1\over 2\pi} {d\over dt} \phi(t)
\end{equation}

- or 

\begin{equation}
\phi(t) = 2\pi k_f \int_{-\infty}^t m(\tau) d\tau
\end{equation}

- $k_f$: frequency deviation constant

## Phase Modulation (PM)
- the phase of the carrier is changed according to the variations in the message signal

\begin{equation}
u(t) = A_c\cos(2\pi f_c t + \phi(t))
\end{equation}

- for PM system,

\begin{equation}
\phi(t) = k_p m(t)
\end{equation}

- $k_p$: phase deviation constant

## Phasor Representation

- signal is represented by a rotating vector in complex plane

\begin{equation}
\phi(t) = A \cos(\theta_i(t)) = A\cos(2\pi f_ct + \gamma(t))
\end{equation}

- $A$: phasor magnitude
- $\theta_i(t)$: instantaneous phase angle (determines position of rotating vector at time $t$)
- $\gamma(t)$: the message signal input

### Angular Velocity (frequency)

<img src="img/Snip20191029_56.png" width=40%/>

\begin{equation}
\textrm{average frequency for some duration: }\omega_{\Delta t}(t) = {\theta_i(t + \Delta t) - \theta_i(t) \over \Delta t} \\
\textrm{instantaneous frequency at time t: } \omega_i(t) = \lim_{\Delta t \rightarrow 0} \omega_{\Delta t}(t) = {d\theta_i(t) \over dt} = \omega_c + {d\gamma(t) \over dt} \\
\textrm{angle: } \theta_i(t) = \int_0^t \omega_i(\tau) d\tau + \theta_0
\end{equation}

# Phase Modulation

<img src="img/Snip20191029_57.png" width=80%/>

- instantaneous angle: $\theta_i(t) = \omega_c t + k_p f(t) + \theta_0$
    - phase $\propto f(t)$ 
- instantaneous frequency: $\omega_i (t) = {d\theta_i(t) \over dt} = \omega_c + k_p {df(t) \over dt}$
- maximum (peak) phase deviation: $\Delta(\theta) = k_p \max(|f(t)|)$


# Frequency Modulation

<img src="img/Snip20191029_58.png" width=80%/>

- instantaneous angle: $\theta_i(t) = \omega_c t + k_{\omega} \int_0^t f(\tau)d\tau + \theta_0$
- instantaneous frequency: $\omega_i (t) = {d\theta_i(t) \over dt} = \omega_c + k_{\omega}f(t)$, so the frequency $\propto f(t)$ 
- maximum (peak) frequency deviation: $\Delta\omega = k_{\omega} \max(|f(t)|)$, $\Delta f = {\Delta \omega \over 2\pi} = k_f \max(|f(t)|)$

## Modulation Index for PM/FM
- PM
\begin{equation}
\beta = \beta_p = \Delta \theta
\end{equation}

- FM
\begin{equation}
\beta = \beta_f = {\Delta \omega \over 2 \pi B} = {k_{\omega} max|f(t)| \over 2 \pi B} = {\Delta f \over B}
\end{equation}
- B (in Hz) is the bandwidth of the message signal

- Narrowband vs Wideband
    - depending on the values of $\beta$, the frequency/phase modulation can be distinguished into two cases
    - narrowband FM/PM when $\beta << 1$
    - wideband FM/PM when $\beta$ is large


## Narrowband FM (NBFM)

\begin{equation}
\Phi(t) = A_c \cos(\omega_c t + \gamma(t)) = A_c \cos(\omega_c t) \cos(\gamma(t)) - A_c \sin(\omega_c t) \sin(\gamma(t))
\end{equation}

- assumptions
    - $|\gamma(t)| << 1$
    - $\cos(\gamma(t)) \approx 1$
    - $\sin(\gamma(t)) \approx \gamma(t)$

\begin{equation}
\textrm{thus, } \Phi_{NBFM}(t) \approx A_c \cos(\omega_c t) - A_c \gamma(t) \sin(\omega_c t)
\end{equation}

# NBFM Single Tone Modulation

- assume $f(t) = a\cos(\omega_m t)$, then single tone modulation

\begin{equation}
\phi(t) = A_c \cos(\omega_c t + k_{\omega} \int_0^t f(\tau)d\tau + \theta_0), \theta_0 = 0
\end{equation}

- instantaneous frequency $\omega_i(t) = {d\theta_i(t) \over dt} = \omega_c + k_{\omega} f(t) = \omega_c + \alpha k_{\omega} \cos(\omega_m t)$
- peak frequency deviation $\Delta\omega = k_{\omega} \max(|f(t)|) = \alpha k_{\omega}$
- instantaneous angle $\theta_i(t) = \omega_c t + k_{\omega} \int_0^t f(\tau)d\tau = \omega_c t + {\alpha k_{\omega} \over \omega_m} \sin(\omega_m t)$
- modulation index of FM signal $\beta = {\Delta\omega \over \omega_m} = {\alpha k_{\omega} \over \omega_m}$

\begin{equation}
\phi(t) = A_c \cos(\omega_c t + \beta \sin(\omega_m t)) \\
= A_c \cos(\omega_c t) \cos(\beta \sin(\omega_m t)) - A_c \sin(\omega_c t) \sin(\beta \sin(\omega_m t))
\end{equation}

- since it is NBFM, then $|\gamma(t)| = |k_{\omega} \int_0^t f(\tau)d\tau| = |\beta \sin(\omega_m t)| << 1$, thus $|\beta| << 1$
- so $\cos(\beta \sin(\omega_m t)) \approx 1$, and $\sin(\beta \sin(\omega_m t)) \approx \beta\sin(\omega_m t)$

\begin{equation}
\phi(t) \approx A_c \cos(\omega_c t) - A_c \beta \sin(\omega_m t)\sin(\omega_c t) \\
\approx A_c \sqrt{1+\beta^2\sin^2(\omega_m t)} \cos(\omega_c t + \tan^{-1}(\beta\sin(\omega_m t)))
\end{equation}

- rearranged as $\phi(t) \approx k \cos(\omega_c t + \alpha)$
- $k = A_c \sqrt{1+\beta^2\sin^2(\omega_m t)}$
    - $\omega_m$ modualates the amplitude, therefore the envelope suffers distortion, unless $\beta$ is very small
- $\alpha = \tan^{-1}(\beta\sin(\omega_m t))$

# Wideband FM (WBFM)

\begin{equation}
\phi(t) = A_c \sum_{n = -\infty}^{\infty} J_n(\beta)\cos((\omega_c + n\omega_m)t)
\end{equation}


# WBFM Single Tone Modulation

- $f(t) = a\cos(\omega_m t)$
- $\phi(t) = A_c \cos(\omega_c t + k_{\omega}\int_0^t f(\tau)d\tau) = A_c\cos(\omega_c t + \beta\sin(\omega_m t))$
- and we want to write the above expression in terms of $\cos(\omega_ct + n\omega_m t)$

\begin{equation}
\phi(t) = A_c Re\{e^{j(\omega_c t + \beta\sin(\omega_m t))}\} = A_c Re\{e^{j\omega_c t} x(t)\} \\
= A_c Re\{e^{j\omega_c t} \sum_{n = -\infty}^{n=\infty} J_n(\beta) e^{jn\omega_m t}\} \\
= A_c \sum_{n = -\infty}^{n=\infty} J_n(\beta)\cos((\omega_c + n\omega_m)t)
\end{equation}

\begin{equation}
\Phi(\omega) = F(\phi(t))
= A_c \pi \sum_{n = -\infty}^{n=\infty} J_n(\beta)(\delta(\omega - \omega_c - n\omega_m) + \delta(\omega+\omega_c+n\omega_m))
\end{equation}

- $J_n(\beta)$: the Bessel function of the first kind, real valued
    - $J_n(\beta) = (-1)^n J_{-n}(\beta)$
    - $J_n(\beta) = \{^{J_{-n}(\beta)\textrm{, if n is even}}_{-J_{-n}(\beta) \textrm{, if n is odd}} $
    - for $\beta << 1$
        - $J_0(\beta) \approx 1$
        - $J_1(\beta) \approx \beta/2$
        - $J_n(\beta) \approx 0$ for $n\ge 2$
    - $\sum_{n = -\infty}^{\infty} J_n^2(\beta) = 1$

<img src="img/Snip20191105_127.png" width=80%/>

## Obtain NBFM from WBFM

- for $\beta << 1$, we have $J_n(\beta) \approx 0$ for $n\ge 2$
- thus $\phi(t) \approx A_c J_0(\beta)\cos(\omega_c t) + A_c J_1(\beta)\cos((\omega_c + \omega_m) t) + A_c J_{-1}(\beta)\cos((\omega_c - \omega_m) t)$
    - $J_0(\beta) \approx 1$
    - $J_1(\beta) \approx \beta/2$
    - $J_{-1}(\beta) \approx -\beta /2$
- $\phi(t) \approx A_c \cos(\omega_c t) - A_c \beta \sin(\omega_m t)\sin(\omega_c t)$

# Bandwidth of FM Signals

\begin{equation}
\phi(t) = A_c \sum_{n = -\infty}^{\infty} J_n(\beta)\cos((\omega_c + n\omega_m)t)
\end{equation}

\begin{equation}
\Phi(\omega) = A_c \pi \sum_{n = -\infty}^{n=\infty} J_n(\beta)(\delta(\omega - \omega_c - n\omega_m) + \delta(\omega+\omega_c+n\omega_m))
\end{equation}

- $\Phi(\omega)$ contains a carrier component and an infinite set of side frequencies located symmetrically on either side of carrier frequency (i.e., $\pm n\omega_m$)
- these frequencies are integer multiples of $\omega_m$, defined as **harmonics**

<img src="img/Snip20191105_128.png"/>

- in general, $B(rad) = 2n\omega_m$, where $n$ depends on defintion of **significant sidebands**
- for $\beta <<1$, only $n = 0, \pm 1$ terms exist, thus $B(rad) = 2\omega_m$


# Significant Sidebands Definition

- A sideband is significant if its magnitude is equal to or exceeds 1\% of the unmodulated carrier, i.e.,
\begin{equation}
|J_n(\beta) \ge 0.01|
\end{equation}
- $J_n(\beta)$ diminishes rapidly and the ratio $n/\beta \rightarrow 1$ as $\beta$ becomes large
\begin{equation}
\beta >> 1, n \approx \beta
\end{equation}

<img src="img/Snip20191105_129.png" width=40%/>

- the bandwidth for large $\beta$ can be approximated as

\begin{equation}
B(rad/s) = 2n\omega_m \approx 2\beta\omega_m = 2\Delta\omega \\
\beta = {\Delta\omega \over \omega_m} = {ak_{\omega} \over \omega_m} = {a2\pi k_f \over 2\pi f_m}
\end{equation}

# Carson's Rule for Bandwidth of FM Signals

<img src="img/carson.jpg" width=500%/>

\begin{equation}
B\approx 2f_m(1+\beta)\\
B\approx 2(f_m +\Delta f)
\end{equation}

- for very small $\beta$, $B \approx 2f_m(1+\beta) \approx 2f_m$
- for very large $\beta$, $B \approx 2f_m(1+\beta) \approx 2f_m\beta =2\Delta f$

- Carson's Rule holds for general modulating signals that are band-limited and have finite power
- Carson's Rule gives less bandwidth than the definition of significant bands
    - not true for the limiting cases
    
**Example** A 10MHz is frequency-modulated by a sinusoidal signal such that the peak frequency deviation is 50kHz. Determine the bandwidth of the FM signal if the frequency of the modulating sinusoid is

- a) 500kHz
    - $\beta = \Delta f/ f_m = 50*10^3 / 500 * 10^3 = 0.1$ -> narrowband signal
    - $B \approx 2f_m = 1MHz$
    - $B = 2f_m(1+\beta) = 1.1MHz$ (Carson's rule)
- b) 500 Hz
    - $\beta = \Delta f / f_m = 100$ -> wideband signal
    - $B \approx 2\Delta f = 100kHz$
    - $B = 2f_m(1+\beta) = 101kHz$ (Carson's rule)
- c) 10 kHz
    - $\beta = \Delta f / f_m = 5$
    - $B = 2nf_m = 2 * 8 * 10 * 10^3= 160kHz$ using the definition of significant bands
    - $B = 2f_m(1+\beta) = 120kHz$ (Carson's rule)

# Average Power of FM Signals

- $\phi(t)= A_c\cos(\omega_c t + \beta\sin(\omega_m t))$
- total average power $P = \bar{\phi^2(t)} = A_c^2/2$
- alternatively: $\phi(t) = A_c \sum_{n = -\infty}^{\infty} J_n(\beta)\cos((\omega_c + n\omega_m)t)$
\begin{equation}
\Phi(\omega) = A_c \pi \sum_{n = -\infty}^{n=\infty} J_n(\beta)(\delta(\omega - \omega_c - n\omega_m) + \delta(\omega+\omega_c+n\omega_m))
\end{equation}

- power spectral density
\begin{equation}
S_{\phi}(\omega) = 2\pi \sum_{n=-\infty}^{\infty} |{A_c J_n(\beta) \over 2}|^2 (\delta(\omega - \omega_c - n\omega_m) + \delta(\omega + \omega_c + n\omega_m))
\end{equation}

- total average power

\begin{equation}
P = {1\over 2\pi} \int_{-\infty}^{\infty} S_{\phi}(\omega)d\omega = 2 {A_c^2 \over 4} \sum_{n=-\infty}^{\infty} J_n^2(\beta) = {A_c^2 / 2}
\end{equation}

**Example** carrier $c(t) = 10\cos(\omega_c t)$, message $f(t) = \cos(20\pi t)$, the message is used to frequency modulate the carrier with $k_{\omega} = 100\pi$, find the expression for the modulated signal and determine how many harmonics should be selected to contain 99% of the modulated signal power
- $\phi(t) = 10\cos(\omega_c t + k_{\omega} \int_0^t \cos(20\pi \tau)d\tau) = 10\cos(\omega_c t + 5\sin(20\pi t))$
- $\Phi(t) = 10\sum_{n=-\infty}^{\infty}J_n(t) \cos((\omega_c + 20\pi n)t)$
- A_c = 10, total average power $P = {A_c^2 \over 2} = 50$ => $99\%P = 49.5$
- $2 {A_c^2 \over 4} (J_0^2(t) + 2\sum_{n=1}^k J_n^2(5)) \ge 49.5$ => trial and error, $k=6$
