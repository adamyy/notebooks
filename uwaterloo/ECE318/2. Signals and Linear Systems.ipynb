## 2.1 Review of Fourier Series and Transform

## Fourier Series Representation for Periodic Signals

**Definition**: Let signal $x(t)$ be a periodic signal with fundamental period $T$ (i.e., the smallest number $T>0$ satisfying $x(t+T)=x(t)$ for all $t$), sastifying the *Dirichlet conditions*. Then $x(t)$ can be expanded in terms of the complex exponential signals $\{e^{j\omega_0nt}\}_{n=-\infty}^{n=\infty}$ as

\begin{equation}
x(t) = \sum_{n=-\infty}^{+\infty}X_ne^{j\omega_0nt}
\end{equation}

\begin{equation}
X_n = {1 \over T} \int_T x(t)e^{-j\omega_0nt} dt
\end{equation}

- The coefficients $X_n$ are called the *Fourier series coefficients* of the signal $x(t)$
- $\omega_0=2\pi/T$ \[rad/s\] are called the *fundamental frequency*, the frequencies of the complex exponential signals are multiples of this frequency
    - the $n$th multiple of the fundamental frequency for postive $n$ is called the $n$th *harmonic*

**Why Fourier**: Fourier is useful in studying LTI systems, complex exponentials are eigenfunctions of LTI

### Aside

For any signal $x(t)$ defined on $[t_1, t_2]$, we can represent it as a linear combination of a set of orthogonal signals $\{\varphi_n(t)\}$
\begin{equation}
x(t) = \sum_n x_n \varphi_n(t), t_1<t<t_2
\end{equation}
where
\begin{equation}
\int_{t_1}^{t_2}\varphi_n(t)\varphi_m^{*}(t)dt=\{^{K_n, n=m}_{0, n \ne m}
\end{equation}

\begin{equation}
x_n={1 \over K_n}\int_{t_1}^{t_2}x(t)\varphi_n^*(t)dt
\end{equation}

The orthogonal signal set $\{\varphi_n(t)\}|_{n=1}^{N(\rightarrow \infty)}$ forms a signal space of N dimensions.

$x(t)$ can be represented by its coordinate ($x_1, x_2, ...$) in the signal space, given $\{\varphi_n(t)\}$. As a result, we can use a constant coordinate to represent a time-varying signal, where the time-varying component is represented by the axes $\{\varphi_n(t)\}$ in the signal space.

**Special case**
- if $x(t)$ is a periodic signal with period $T$, then the representation is valid over every period


## Alternative Representation for Real Periodic Signal: the Trigonometric Fourier Series

\begin{equation}
X_n = {1 \over T} \int_T x(t)e^{-j\omega_0nt}dt = A_n - j\ B_n = {1 \over T} \int_T {x(t)\cos(\omega_0nt)dt} - j {1 \over T} \int_T {x(t)\sin(\omega_0nt)dt}
\end{equation}

\begin{equation}
|X_n|=\sqrt {A_n^2+B_n^2}, \theta_n=-\arctan{B_n \over A_n}
\end{equation}

**Representation**
\begin{equation}
x(t) = X_0 + 2 \sum_{n=1}^{\infty}|X_n|\cos(\omega_0nt+\theta_n) 
\end{equation}
where 
\begin{equation}
X_0 = {1 \over T} \int_T x(t)dt
\end{equation}

**Proof**

\begin{equation}
x(t)=\sum_{n=-\infty}^{+\infty}X_ne^{j\omega_0nt}
= X_0+\sum_{n=1}^{+\infty}(X_ne^{j\omega_0nt}+X_{-n}e^{-j\omega_0nt})
= X_0+2\sum_{n=1}^{\infty}Re(X_ne^{j\omega_0nt})
\end{equation}

\begin{equation}
X_n={1 \over T} \int_T x(t)e^{-j\omega_0nt}dt = A_n - j\ B_n =
{1 \over T}\int_T {x(t)\cos(\omega_0nt)dt} - j\ {1 \over T}\int_T {x(t)\sin(\omega_0nt)dt} 
\end{equation}

\begin{equation}
x(t) = X_0 + 2 \sum_{n=1}^{\infty}(A_n\cos(\omega_0nt)+B_n\sin(\omega_0nt))
\end{equation}

- if $x(t)$ is an even function, i.e., $x(-t) = x(t)$, then $B_n=0$
- if $x(t)$ is an odd function, i.e, $x(-t) = -x(t)$, then $A_n=0$


## Fourier Spectrum

- Time-domain representation
\begin{equation}
x(t) = \sum_n X_n e^{jn\omega_0t} = \sum_n |X_n| e^{j(n\omega_0t+\theta_n)}
\end{equation}
where
\begin{equation}
X_n = |X_n|e^{j\angle X_n}
\end{equation}

- Frequency-domain representation
\begin{equation}
\angle X_n \sim \omega, phase\ spectrum
\end{equation}
\begin{equation}
|X_n| \sim \omega, magnitude\ spectrum
\end{equation}

**Summary**: The Fourier series is a means for expanding an "well-behaved" signal over $[t_1, t_2]$ or a periodic signal in terms of orthogonal complex exponentials. The expansion in terms of complex exponentials is particularly useful when analyzing LTI systems.

**Example**: Determine the Fourier series expansion for the Periodic Gate Function
\begin{equation}
x(t) = A \sum_{n=-\infty}^{n=\infty} rect({{t - nT} \over \tau})
\end{equation}

where

\begin{equation}
rect({t\over \tau}) = \{_{0, |t| \geq \tau/2}^{1, |t| < \tau/2}
\end{equation}

**Solution**
- $x(t)$ is an even periodic function $x(t)=x(-t)$, therefore $B_n=0$
- therefore

\begin{equation}
{when\ n \ne 0}, X_n={1 \over T} \int_T x(t)cos(\omega_0nt)dt
= {A \over T} \int_{-\tau/2}^{\tau/2} \cos(\omega_0nt)dt
= {A \over T} {2 \over \omega_0n} \sin(\omega_0 n {\tau \over 2}) 
\end{equation}

\begin{equation}
X_0 = {1 \over T} \int_T x(t)dt = {A \over T} \int_{-\tau/2}^{\tau/2}dt = {A\tau \over T}
\end{equation}

**Definition**: *sine over argument* function $sa(x)={\sin(x) \over x}$

Therefore,

\begin{equation}
X_n = {A \over T} {2 \over \omega_0n} \sin(\omega_0 n {\tau \over 2}) 
= {A\tau \over T} sa({\omega_0n\tau \over 2}) 
= {A\tau \over T} sa(n\pi {\tau \over T})
\end{equation}

- the amplitude decreases proportional to $1/T$
- the spacing between lines decreases proportional to $1/T$
- zero crossings of the spectrum remain the same and not dependent on T

## Fourier Transform

- the Fourier series is a means for expanding a periodic signal in terms of complex exponentials
- for an aperiodic function, we also want to express it as a sum of exponential signals
- construct a new periodic function $x_T(t)$ based on the original aperiodic signal $x(t)$ such that $\lim_{T\rightarrow \infty} x_T(t)=x(t)$
- then $x_T(t)$ is a periodic signal and can be represented by Fourier series
\begin{equation}
x_T(t)= \sum_{n=-\infty}^{+\infty}X_ne^{j\omega_0nt}, \textrm{ where } X_n = {1 \over T} \int_T x_T(t) e^{-j\omega_0nt}dt
\end{equation}

define
\begin{equation}
\omega_n = n \omega_0, X(\omega_n) = TX_n
\end{equation}

then
\begin{equation}
X(\omega_n) = \int_T x_T(t)e^{-j\omega_nt}dt, \textrm{ and } x_T(t) = \sum_{n=-\infty}^{+\infty} {1 \over T} X(\omega_n)e^{j\omega_nt}
\end{equation}

defining $\Delta \omega = 2\pi / T$, then
\begin{equation}
x_T(t) = {1 \over 2\pi} \sum_{n=-\infty}^{+\infty}X(\omega_n)e^{j\omega_n t} \Delta \omega
\end{equation}

as $T becomes large, $\Delta \omega$ becomes smaller, take limit for $T \rightarrow \infty$ on both sides:
\begin{equation}
\lim_{T\rightarrow \infty} x_T(t) = x(t) = \lim_{T\rightarrow \infty} {1 \over 2\pi} \sum_{n=-\infty}^{+\infty}X(\omega_n)e^{j\omega_n t} \Delta \omega
\end{equation}

thus:
\begin{equation}
x(t) = {1 \over 2\pi} \int_{-\infty}^{+\infty} X(\omega) e^{j\omega t}
\end{equation}


### Fourier Transform

\begin{equation}
X(\omega) = \mathcal{F}\{x(t)\} = \int_{-\infty}^{+\infty}x(t)e^{-j\omega t}dt
\end{equation}

$X(\omega)$ also known as *spectral-density function* of $x(t)$

### Inverse Fourier Transform

\begin{equation}
x(t) = \mathcal{F}^{-1}\{X(\omega)\} = {1 \over 2\pi} \int_{-\infty}^{+\infty} X(\omega)e^{j\omega t} d\omega
\end{equation}

### Relationship between Fourier Series & Fourier Transform

\begin{equation}
X_n = {1 \over T} X(\omega) |_{\omega \rightarrow n\omega_0}
\end{equation}

\begin{equation}
X(\omega) = TX_n |_{\omega_0 \rightarrow {\omega \over n}}
\end{equation}

**Example** Determine the Fourier transform of Gate function

\begin{equation}
\textrm{rect}({t \over \tau}) = \{_{0, |t|\geq \tau/2}^{1, |t| < \tau /2}
\end{equation}

<img src="img/Snip20190917_90.png" width=30%/>

- $x(t)$ is an even aperiodic function
- $X(\omega) = \int_{-\infty}^{+\infty}x(t)\cos(\omega t) dt = A \int_{-\tau / 2}^{\tau / 2} \cos(\omega t) dt = {2A \over \omega} \sin(\omega {\tau \over 2}) = A\tau sa({\omega \tau \over 2})$
- alternative solution, if we know the Fourier series for a periodic gate function is $X_n = {A\tau \over T} sa({\omega_0 n \tau \over 2})$
- then by the relation to Fourier series, we have $X(\omega) = TX_n|_{\omega_0 \rightarrow {\omega \over n}}$
- $X(\omega) = T {A\tau \over T} sa({\omega_0 n \tau \over 2})|_{\omega_0 \rightarrow {\omega \over n}} = A\tau sa({\omega \tau \over 2})$

**Observation** Compared to the spectrum to periodic signals, aperiodic signals have continuous spectrum


## Relation of Fourier Transform in $\omega$ and $f$

If the variable in the Fourier transform is chosen to be $f=\omega/{2\pi}$ (i.e., frequency in Hz) rather than $\omega$, then we have

**Fourier Transform** $X(f)=\mathcal{F}\{x(t)\} = \int_{-\infty}^{+\infty}x(t)e^{-j2\pi ft}dt$

**Inverse Fourier Transform** $x(t) = \mathcal{F}^{-1}\{X(f)\} = \int_{-\infty}^{+\infty}X(f)e^{j2\pi f t} df$


## Fourier Transform for Periodic Signals

- periodic signal can be represented by its exponential Fourier series

\begin{equation}
x(t) = \sum_{n=-\infty}^{+\infty}X_ne^{j\omega_0nt} \textrm{, } \omega_0 = 2\pi /T
\end{equation}

- the Fourier transform of a periodic signal consists of a set of impulses located at the multiples of the fundamental frequency

\begin{equation}
\mathcal{F}\{x(t)\}=\mathcal{F}\{\sum_{n=-\infty}^{+\infty}X_ne^{j\omega_0nt}\}=\sum_{n=-\infty}^{+\infty}X_n\mathcal{F}\{e^{j\omega_0 nt}\} = 2\pi \sum_{n=-\infty}^{+\infty} X_n \delta(\omega - n\omega_0)
\end{equation}



# Properties of the Fourier Transform

## Linearity (Superposition)
\begin{equation}
ax(t)+by(t) \iff aX(\omega)+bY(\omega)
\end{equation}

**Proof** $\mathcal{F}\{ax(t)+by(t)\} = \int_{-\infty}^{+\infty} [ax(t)+by(t)]e^{-j\omega t} dt = a \int_{-\infty}^{+\infty} x(t)e^{-j\omega t} dt + b \int_{-\infty}^{+\infty} y(t)e^{-j\omega t} dt = aX(\omega) + bY(\omega)$

## Comple Conjugate

\begin{equation}
x(t) \iff X(\omega) \textrm{,  } x^*(t) \iff X^*(-\omega)
\end{equation}

## Duality

\begin{equation}
\mathcal{F}\{X(t)\} = 2\pi x(-\omega)
\end{equation}

\begin{equation}
\mathcal{F}\{X(-t)\} = 2\pi x(\omega)
\end{equation}

**Proof**

\begin{equation}
x(t) = {1 \over 2\pi} \int_{-\infty}^{+\infty} X(\omega) e^{j\omega t} d\omega
\end{equation}

Let $\omega = -u$,
\begin{equation}
x(t) = {1 \over 2\pi} \int_{-\infty}^{+\infty} X(-u) e^{-jut} du
\end{equation}

Let $t = \omega$
\begin{equation}
x(\omega) = {1 \over 2\pi} \int_{-\infty}^{+\infty} X(-u) e^{-ju\omega} du = {1 \over 2\pi} \int_{-\infty}^{+\infty} X(-t) e^{-j\omega t} dt = {1\over 2\pi} \mathcal{F}\{X(-t)\}
\end{equation}

## Coordinate Scaling
\begin{equation}
