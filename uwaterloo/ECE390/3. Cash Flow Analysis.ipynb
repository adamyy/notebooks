# 3.3 Compound Interest Factors for Discrete Compounding

- **Compound interest factors** are formulas that define mathematical equivalence for specific common cash flow patterns

- the four discrete cash flow patterns that are commonly used to model the timing of receipts and disbursements are
    - 1. a single disbursement or receipt
    - 2. a set of equal disbursements or receipts over a sequence of periods, referred to as an **annuity**
    - 3. a set of disbursements or receipts that change by a constant *amount* from one period to the next in a sequence of periods, referred to as an **arithmetic gradient series**
    - 4. a set of disbursements or receipts that change by a constant *proportion* from one period to the next in a sequence of periods, referred to as a **geometric gradient series**

# 3.4 Compound Interest Factors for Single Disbursements or Receipts

<img src="img/Snip20191009_145.png" width=80%/>

- two commonly used factors relate a single cash flow in one period to another single cash flow in a different period are **compound amount factor** and the **present worth factor**


- ***Q: “How much do I have in the end ($F$), given I have this much money ($P$), and interest rate is $i$ over $N$ number of periods”***
- *compound amount factor*: $(F/P,i,N) = (1+i)^N$
    - gives the future amount (or final value) $F$ that is equivalent to the present amount $P$, when the interest rate is $i$ and the number of periods is $N$
    - relates present and final values: $F = P\times(F/P, i, N) = P \times (1+i)^N$


- ***Q: "if someone are to give me $F$ amount of money in $N$ periods, given interest rate of $i$, how much is that 'future money' worth right now?"***
- *present worth factor*: $(P/F, i, N) = {1 \over (1+i)^N}$
    - gives the current worth $P$ that is equivalent to a future amount $F$ when the interest rate is $i$ and the number of periods is $N$


# 3.5 Compound Interest Factors for Annuities

<img src="img/Snip20191009_146.png" width=80%/>

- a series of uniform receipts or disbursements that start at the end of the first period and continue over $N$ periods


- ***Q: "if someone are to give me $A$ amount of money for every period in the next $N$ periods, given interest rate $i$, how much do I have in the end?"***
    - *sinking fund factor*: $(A/F, i, N) = {i \over (1+i)^N - 1}$
        - $A$: a repeated receipt or disbursement
        - $F$: the future amount
        - $i$: interest rate
        - $N$: number of periods
        - the name comes from the concept of *sinking fund*: an interest-baring account into which regular depositis are made in order to accumulate some amount
    

- ***Q: "if I want to have $F$ amount of money in the end of the $N$th period, given interest rate of $i$, how much should I deposit every period?"***
    - *uniform series compound amount factor*: $(F/A, i, N) = {(1+i)^N - 1 \over i}$

    
- ***Q: "if I spend $P$ amount of money right now on some project that gives me periodic return, given interest rate $i$, and the project's service life is $N$ periods, how much should I earn every period to break even in the end?"***
    - *capital recovery factor*: $(A/P, i, N) = {i(1+i)^N \over (1+i)^N - 1}$
    
    
- ***Q: "if someone are to promise to give me $A$ amount every period for the next $N$ periods, given interest rate of $i$, how much money is that worth right now?"***
    - *series present worth factor*: $(P/A, i, N) = {(1+i)^N - 1 \over i(1+i)^N}$


# 3.6 Arithmetic Gradient Series

<img src="img/Snip20190917_87.png" width=80%/>

- a series of receipts or disbursements that starts at zero at the end of the first period and then increase by a constant amount from period to period
- e.g., we may model a pattern of increasing operating costs for an againg machine as an arithmetic gradient series if the costs are increasing by approximately the same amount each period
- **arithmetic gradient to annuity conversion factor**: $(A/G, i, N) = {1 \over i} - {N \over (1+i)^N - 1} = {A\over G}$
- there is often a base annuity $A'$ associated with a gradient, to determine the uniform series equivalent to the *total* cash flow, the base annuity $A'$ must be included to give the overall annuity
    - $A_{tot} = A' + G(A/G, i, N)$

<img src="img/Snip20190917_88.png" width=60%/>

**Example**: 650 MW combinational cycle, capital costs \\$948/kW, fixed O&M cost \\$14.39/kW, variable O&M cost \\$3.43/MW, capacity factor 11.4\%, N = 20, discount rate 8\%, average HOEP is approximately \\$60/mWh, assume that capacity factor grows linearly 10\% ~ 15\% in 20 years. Find $P$.

<img src="img/Snip20190917_89.png" width=80%/>

- Income: $A'_m = 10\% * 650MW * 365 d * 24 h/d * 60 \$/MWh = \$34,164.00$
- $(N-1)G_{in} = 5\% * 650 MW * 365 d * 24 h/d * 60 \$/MWh$, so $G_{in} = \$899,053$
- $A_{T_{in}} = A'_{in} + G_{in}(A_{in}/G_{in}, i, N) = A'_{in} + G_{in}({1 \over i} - {N \over (1+i)^N - 1}) = \$40,490,589$
- $P_{in} = A_{T_{in}}(P_{in}/A_{T_{in}}, i, N) = A_{T_{in}} {(1+i)^N - 1 \over i+(1+i)^N} = \$399,542,572$
- O&M Cost: $A_{tot} = A'_{tot} + G_{out}({1 \over i} - {N \over (1+i)^N - 1})$





# 3.7 Geometric Gradient Series


- a series of cash flows that increase or decrease by a constant *percentage* each period
- may be used to model **inflation or deflation**, *productivity improvement or degradation*, *growth or shrinkage of market size*, etc.
- the base value of the series is $A$
- the **growth** rate is referred to as $g$
    - the terms in such a series are given by $A, A(1+g), A(1+g)^2, ..., A(1+g)^{N-1}$ at the ends of periods $1,2,3,...,N$ respectively


<img src="img/Snip20190919_112.png" width=80%/>
<img src="img/Snip20190919_113.png" width=80%/>

- **Geometric Gradient To Present Worth Conversion Factor**: $(P/A, g, i, N)$
    - present worth: $P$
    - number of periods: $N$
    - base receipt or disbursement: $A$
    - rate of growth $g$
    - interest rate $i$

\begin{equation}
P = {A \over 1+i} + {A(1+g) \over (1+i)^2} +...+ {A(1+g)^{N-1} \over (1+i)^N}
\end{equation}

- **Growth-adjusted Interest Rate** $i^o = {1+i \over 1+g} - 1$
    - $(P/A, g, i, N) = {(P/A, i^o, N) \over 1 + g}$
    - $(P/A, g, i, N) = {(1 + i^o)^N - 1 \over i^o (1+i^o)^N} {1 \over 1 + g}$

1. $i>g>0$, **growth is positive, but less than the rate of interest**, in which case the growth-adjusted interest rate $i^o$ is positive
2. $g>i>0$, **growth is positive and greater than the interest rate**, the growth-adjusted interest rate $i^o$ rate is negative
3. $g = i > 0$, **growth is positive and exactly equal to the interest rate**, the growth-adjusted interest rate $i^o$ is zero
4. $g<0$, **Growth is negative**, the series is decreasing and the GAIR equals zero

**Example** Retirement funds, A = 100 dollars per month, N' = 12 months, interest rate $i = 5\%$, $g = 3\%$ inflation, $N = 20 years$


# 3.8 Non-Standard Annuities and Gradients

Standard assumption for annuities and gradients is that the payment period and compounding period are the same. If they are not, there are three methods for dealing with this situation

1. treat each cash flow in the annuity or gradient individually
    - most useful when the annuity or gradient series is not large
2. convert the non-standard annuity or gradient to standard form by changing the compounding period
3. convert the non-standard annuity to standard form by finding an equivalent standard annuity for the compounding period, cannot be used for gradients



# 3.9 Present Worth Computations When $N\rightarrow \infty$

The present worth of an infinitely long uniform series of cash flows is called the **capitalized value** of the series.

\begin{equation}
P=\lim_{N\rightarrow \infty} A(P/A, i, N) = A\lim_{N\rightarrow \infty} [{(1 +i)^N - 1 \over i(1+i)^N}] = A\lim_{N\rightarrow \infty} [{1 - {1 \over (1+i)^N} \over i}] = A/i
\end{equation}

# Derivation of Discrete Compound Interest Factors

$F=P(1+i)^N$

## Compound Amount Factor

\begin{equation}
F/P = (F/P, i, N) = (1+i)^N
\end{equation}

## Present Worth Factor
- converts a future amount $F$ to a present amount $P$

\begin{equation}
P = F(P/F, i, N) \rightarrow (P/F, i, N) = {1 \over (1+i)^N}
\end{equation}


## Sinking Fund Factor
- if a series of payments $A$ follows the pattern of a standard annuity of $N$ payments in length, then the future value of the payment in the $j$th period is

\begin{equation}
F=A(1+i)^{N-j}
\end{equation}

- the future value of all the annuity payments is then
\begin{equation}
F = A(1+i)^{N-1} + A(1+i)^{N-2} + ... + A(1+i) + A = A((1+i)^{N-1} + (1+i)^{N-2} + ... + (1+i) + 1)
\end{equation}

- We are trying to find the ratio of A to F, then 
\begin{equation}
A = F ({1 \over (1+i)^N - 1})
\end{equation}

- the sinking fund factor is then given by
\begin{equation}
(A/F, i, N) = {i \over (1+i)^N - 1}
\end{equation}

## Uniform Series Compound Amount Factor

- converts an annuity $A$ into a future amount $F$
- take the reciprocal of the sinking fund factor

\begin{equation}
(F/A, i, N) = {(1+i)^N - 1 \over i}
\end{equation}

## Capital Recovery Factor

- if a series of payments $A$ follows the pattern of a standard annuity of $N$ payments in length, then the present value of the payment in the $j$th period is 
\begin{equation}
P_j = A {1 \over (1+i)^j}
\end{equation}

- The present value of the total of all the annuity payments is 

\begin{equation}
P = A({1 \over (1+i)}) + A({1 \over (1+i)^2}) + ... + A({1 \over (1+i)^{N-1}}) + A({1 \over (1+i)^N})
\end{equation}

- The capital recovery factor is given by 

\begin{equation}
A/P = (A/P, i, N) = {i(1+i)^N \over (1+i)^N - 1}
\end{equation}

## Series Present Worth Factor

- converts an annuity $A$ into a present amount $P$

\begin{equation}
P/A = (P/A, i, N) = {(1+i)^N - 1 \over i (1+i)^N}
\end{equation}
