10.	**What is one-dimensional parity checking? Explain its advantage(s) and limitation(s).**
    - Also called single-bit parity checking. It adds a single bit to the data to make the number of 1's even in even parity, or odd in odd parity. After the message is received, the number of 1's is examined to check whether if it matches the parity pattern.
    - Its advantage is that it has very little overhead, with just one redundant EDC bit.
    - Its disadvantage is that it can only detect single bit error.


11.	**What is two-dimensional parity checking? Explain its advantage(s) and limitation(s).**
    - In two-dimensional parity checking, the data is organized in a matrix and parity bits are computed for both rows and columns. This method can both detect and correct single-bit error. However, its disadvantage lies in that it requires a lot of redundant EDC bits, leading to more overhead.


12.	**Explain the idea of Cyclic Redundancy Check (CRC)? Assume a block of data bits, a block of generator bits, and calculate the corresponding CRC bits.**
    - In CRC, a generator $G$ of size $r+1$ is chosen. The CRC bit of $r$ bits is calculated as the remainder of modulo 2 division of $D$ by $G$. CRC method can detect all burst errors less than $r+1$ bits. If $<D',R'>$ received by the receiver gives a non-zero remainder when modolo 2 divided by $G$, then it contains errors.

17.	**Where does frame (packet) collision occur? Why is it important for a transmitters to detect collision?**

23.	**Explain the pure Aloha protocol in brief.**
    - for each node, when a frame arrives, it transmits immediately
    - while the node is transmitting, it listen to the channel and detect collisions
        - if no collisions are detected during the transmission of frame, it concludes that the frame is succcessfully transmitted and transmit the next frame


24.	**Explain the slotted Aloha protocol in brief.**
    - slotted Aloha makes a few assumptions
        - all frames are the same size
        - time is divided into equal size slots
        - nodes only transmit at the slot beginning
        - nodes are synchronized
        - collisions can be detected by each node
    - when a node obtains a fresh frame, it transmits in the next slot
        - if there is no collision in the transmission, the node can send new frame in the next time slot
        - if there is collision, node retransmits frame in each subsequent slot with probability $p$ until success


25.	**How does slotted Aloha differ from pure Aloha?**
    - slotted Aloha divide up the time into equal size time slots and assume that all nodes are synchronized and only transmit at the slot beginning
    - slotted Aloha also assume that all frames are the same size

37.	**What does the Address Resolution Protocol (ARP) do?**

38.	**Briefly explain the ARP protocol.**
    - every node stores an ARP table that maps an IP address to a MAC address, each row contains the IP address, its matching MAC address, and a TTL (time to live, time after which the address mapping will be forgotten)
    - When A wants to send datagram to B, it first look up in its ARP table
        - if the MAC address is found, node A sends the datagram with B's MAC and IP address
        - if the MAC address is not found, node A broadcasts an ARP query packet with B's IP address

54.	**What is the handshake mode of operation of a WLAN? What is its advantage? What is its overhead?**
    - The sender and receiver perform "handshaking" prior to data transfer.
        - The sender sends an RTS frame to receiver, indicating that it wants to transfer data
        - The receiver responds with a CTS frame
        - The sender then proceeds to send the DATA frame
        - After received the data, the receiver replies with an ACK frame
    - Whether or not handshaking mode is used depends on the size of the DATA frame, if it is larger than dotRTSThreshold, then handshaking is used
    - The overhead is the RTS and CTS signal
    - Its reduces the probability of collisions for larger frames

59.	**Explain the hidden terminal problem.**
    - In wireless networking, the hidden terminal problem occurs when a node can communicate with a wireless AP but cannot directly communicate with other nodes that are communicating with that AP.
    - Two nodes A and C are trying to send a message to node B, in which case B can see both nodes but A and C cannot see each other.
    - When both nodes are trying to transmit, collision happens at node B
    - CSMA/CA is a solution to this problem.


60.	**Explain the exposed terminal problem.**
    - In wireless networking, the exposed terminal problem occurs when a node is prevented from sending packets to other nodes because of co-channel interference with a neighboring transmitter. This leads to loss of opportunity to transmit and loss of bandwidth.
    - In an example, say A, B, D are in close proximity and node A, B, C are in close proximity. Node A wants to transfer to node D and node B wants to transfer to node C. In this case, if B transmits a frame to C, it will not collide with the transmission from A to D. However, since B knows someone is transmitting, it will not initiate the transmission because it is unaware of D's location.
    - The problem is due to B being exposed to A's transmission.


61.	**What is carrier sensing and what is virtual carrier sensing in WLAN?**
    - carrier sensing: the process of sensing the medium to see if it is currently being used before attempting to use it
    - virtual carrier sensing (a MAC protocol concept): 
        - an integer variable in each node, called NAV (Network Allocation Vector), which indicates whether or not a nearby hidden node is likely to be transmitting now
            - NAV > 0: most likely another node is transmitting
            - NAV = 0: no one nearby is transmitting
            - NAV < 0: bug case, should not exist
        - transmit condition: medium is idle at the receiver => NAV = 0 and carrier is absent


62.	**How is virtual carrier sensing implemented in WLAN?**

63.	**What is NAV (Network Allocation Vector) and how is a NAV incremented and decremented?**
    - NAV is an integer counter used in VCS
    - if a frame with a *duration* field in its header is recceived by the node, the node update its NAV with the max of the old NAV and the duration field in the frame
    - with each passing microsecond, the NAV is decremented

65.	**What are the four timing intervals in WLAN? How are they related in terms of length?**
    - SIFS (physical level)
        - Short Inter-Frame Space, waiting time before transmitting any two consecutive frames of any type, 10 µs
        - shortest, ensure that an ongoing cycle of Tx (with or without handshake) is not disturbed
    - aSlot (physical level)
        - propagation delay plus some other hardware delay, chosen such that a node can determine if another node initiated a Tx aSlot time before, 20 µs
    - PIFS (MAC level)
        - Priority (in PCF) IFS, equals to SIFS + aSlot
    - DIFS (MAC level)
        - Distributed IFS, equals to PIFS + aSlot
        - DIFS > PIFS enables an AP to become the controller of a BSS
