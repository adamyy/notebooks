5. **What is circuit switching? Explain its advantages and disadvantages.**
    - In circuit-switched networks, the resources needed along a path (buffers, link transmission rate) to provide for communication between the end systems are **reserved** for the duration of the communication session between end systems.


6. **How is circuit switching implemented?**
    - A circuit in a link is implemented with either **frequency-division multiplexing (FDM)** or **time-division multiplexing (TDM)**.
        - with **FDM**, the frequency spectrum of a link is divided up among the connections established across the link, and the link dedicates a frequency band to each connection for the duration of the connection
        -  with **TDM**, time is divided into frames of fixed duration, and each frame is divided into a fixed number of time slots; when the network establishes a connection across a link, the network dedicates one time slot in every frame to this connection

7. **What is meant by "circuit-like behavior" of a computer network?**
    - guaranteed (constant) performance in transmission rate and delay


8. **What is meant by resource reservation in a circuit switched network?**
    - The dedicated circuits are idle during *silent periods*, when one connection is idle and not actively using the link for transmission, its resources (frequency bands or time slots in the links along the connection's route) cannot be used by other ongoing connections.


9. **Name some key resources in a circuit switched network.**
    - frequency bands in FDM (in the links along the connection's route)
    - time slots in TDM (in the links along the connection's route)


10. **What is packet switching? Explain its advantages and disadvantages.**
    - To send a message from a source end system to a destination end system, the source breaks long messages into smaller chunks of data known as **packets**
    - between source and destination, each packet travels through communication links and **packet switches** (routers and link-layer switches)
    - packets are transmitted over each communication link at a rate equal to the *full* transmission rate of the link ($t_{trans}=L/R$)


11. **Compare circuit switching with packet switching.**
    - Circuit switching can guarantee constant transmission rate (performance), but the fact that it reserves resources for connections even though the reserved resource can be sometimes idle means that the link support less users.
    - Packet switching allows more users to use the network because it does not require resources to be reserved for connections. Packet swicthing also allows for bursty data as resources are shared among all connections and it means one connection can leverage more resources when it needs. However, it gives rise to potential excessive congestion that lead to packet delay and loss, which is then needed to be resolved with protocols that provide reliable data transfer and congestion control.


12. **Explain the concept of "store-and-forward" in packet switching.**
    - The packet switch must receive the entire packet before it can begin transmit the first bit of the packet onto the outbound link.

18. **Identify and explain four sources of packet delay in a packet switched network.**
    - Processing Delay: the time required to examine the packet's header and determine where to direct the packet + the time needed to check for bit-level errors in the packet + etc.
    - Queuing Delay: the packet waits to be transmiitted onto the link, the duration of which depends on the queue intensity
    - Transmission Delay: $L/R$, the amount of time required to push all of the packet's bits into the link
    - Propagation Delay: the time required to propagate from the beginning of the link to the next router


19. **Out of the four sources of packet delays, what are the most  deterministic and what are highly variable?**
    - the most determinitsic: processing delay
    - highly variable: queuing delay

30. **Briefly explain the Internet protocol stack**
    - the Internet protocol stack is composed of 5 layers, each layer implement a service via its own internal-layer actions, relying on services provided by layer below
        - application: supporting network applications (software), e.g., FTP, SMTP, HTTP
        - transport: process-process data transfer (software), e.g., TCP, UDP
        - network: routing of datagrams from source to destination (software & hardware), e.g., IP, routing protocols
        - link: data transfer between neighboring network elements (hardware), e.g., Ethernet, WiFi
        - physical: bits on the wire (hardware)
